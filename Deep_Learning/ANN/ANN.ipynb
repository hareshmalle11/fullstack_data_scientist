```python
import pandas as pd
df=pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# select four features
x=df[['CreditScore', 'Age','Balance','EstimatedSalary']]
y=df['Exited']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=keras.Sequential([
    keras.layers.Dense(16,activation='relu',input_shape=(4,)),
    keras.layers.Dense(8,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(
    x_train,y_train,
    epochs=50,
    validation_split=0.2,
    batch_size=32
)
```

```python
accuracy,loss=model.evaluate(x_test,y_test)
print(accuracy,loss)
```

```python
prediction=model.predict(x_test)
predictions=(prediction>0.5)
print(predictions[:10])
```

```python
import numpy as np
```

```python
# remove less co related rows by intailize threshold
threshold=0.4
corr_matrix=df2.corr().abs()
upper=corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(bool))
to_drop=[column for column in upper.columns if any(upper[column]>threshold)]
print(to_drop)
```

```python
# list input and output
x=df2.drop(['Cover_Type','Id','Soil_Type7','Soil_Type15'],axis=1)
y=df2['Cover_Type']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model=keras.Sequential([
    keras.layers.Dense(256,activation='relu',input_shape=(52,)),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dense(16,activation='relu'),
    keras.layers.Dense(8,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history=model.fit(x_train,y_train,epochs=50,validation_split=0.2,batch_size=32)
```

```python
acc,loss=model.evaluate(x_test,y_test)
print(acc,loss)
```
