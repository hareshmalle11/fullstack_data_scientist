```python
import ctypes
import platform
import sys

print("Python:", sys.version)
print("Platform:", platform.platform())
print("Machine:", platform.machine())

def check_dll(name):
    try:
        ctypes.CDLL(name)
        print(f"{name}: OK")
    except OSError as exc:
        print(f"{name}: MISSING ({exc})")

for dll in ["vcruntime140.dll", "vcruntime140_1.dll", "msvcp140.dll"]:
    check_dll(dll)
```
