# Backpropagation, Chain Rule, and Weight Update Process

## 8. Backpropagation (The Core of Neural Network Learning)

### 8.1 What is Backpropagation?

Backpropagation is the algorithm that computes how much each weight contributes to the error (loss). It allows us to update weights in the direction that reduces error.

**Simple Analogy:**

```
Imagine a company with multiple departments:
1. Production (input processing)
2. Assembly (hidden layers)
3. Quality Check (output)
4. Customer feedback received (loss calculated)

If product quality is poor (high loss):
- Quality team identifies what's wrong
- Reports back through Assembly
- Reports back to Production
- Each department understands their contribution to the problem
- Each department makes corrections

Backpropagation: Error flows backward through network
Each layer understands: "How much did I contribute to this error?"
Each layer adjusts to reduce its contribution
```

**Technical Definition:**

Backpropagation is the process of:
1. Computing loss (forward pass)
2. Computing gradients by propagating loss backward through network
3. Using chain rule to calculate тИВLoss/тИВw for each weight
4. Using these gradients to update weights

### 8.2 Why Backpropagation Is Needed

**The Core Problem:**

```
We want to minimize loss by adjusting weights.
But: Loss depends on all weights in complex way!

Example: Network with 1000 weights
Loss = f(wтВБ, wтВВ, ..., wтВБтВАтВАтВА)

Question: How much does loss change if I change wтВЕтВВтВГ?
Answer: Must trace through entire network!
- wтВЕтВВтВГ affects its neuron's output
- That output affects next layer's neurons
- Those affect next layer...
- Eventually affects final output
- Final output affects loss

Without systematic way to trace this:
- Cannot compute gradients
- Cannot train network!

Backpropagation: Systematic way to compute all gradients efficiently!
```

**What It Enables:**

```
1. Gradient Computation
   For every weight: compute тИВLoss/тИВw

2. Efficient Calculation
   Uses chain rule to avoid redundant calculations
   Reuses intermediate results
   O(n) complexity instead of O(n┬▓) or worse

3. Practical Training
   Makes neural networks trainable
   Without backpropagation: optimization impractical
   With backpropagation: feasible, efficient

4. Deep Learning
   Enables training of deep networks
   Each layer computes gradient efficiently
   Propagates back through all layers
```

### 8.3 Goal of Backpropagation

**Primary Goal: Compute Gradients**

```
For each weight w_i in network:
Compute: тИВLoss/тИВw_i

This gradient tells us:
- How much does loss change if increase w_i?
- Which direction to adjust w_i to reduce loss?
- How fast should we change w_i?

Final goal: Use gradients to update weights
w_i_new = w_i - learning_rate ├Ч (тИВLoss/тИВw_i)
```

**Secondary Goals:**

```
1. Reduce Error
   тИВLoss/тИВw tells direction to reduce loss
   Moving opposite to gradient = lower loss

2. Update All Weights Simultaneously
   Compute gradients for all weights at once
   Update all weights based on these gradients
   Coherent learning across network

3. Learn Representations
   Through weight updates, network learns:
   - Which features matter
   - How to transform input toward output
   - Hierarchical representations (in deep networks)
```

### 8.4 Backpropagation Flow

**Step-by-Step Process:**

```
Training Loop:
1. FORWARD PASS
   Input тЖТ Network тЖТ Prediction
   Compute activations of all neurons
   Result: ┼╖ (predicted output)

2. COMPUTE LOSS
   Loss = f(y_actual, ┼╖_predicted)
   Single number: how wrong are we?

3. BACKWARD PASS (Backpropagation)
   Start at output layer
   Compute how much each output contributes to loss
   Propagate back through layers
   Each layer computes: "How much did I cause the error?"

4. COMPUTE GRADIENTS
   From backward pass, get: тИВLoss/тИВw for each weight

5. UPDATE WEIGHTS
   w_new = w_old - lr ├Ч тИВLoss/тИВw

6. REPEAT
   Next batch: back to step 1
```

## 9. Chain Rule (Mathematical Foundation)

Chain rule is the mathematical principle that makes backpropagation possible. It's how we efficiently compute gradients through multiple layers.

### 9.1 What is Chain Rule?

**Mathematical Statement:**

```
If y = f(u) and u = g(x):
Then: dy/dx = (dy/du) ├Ч (du/dx)

Interpretation:
To find how y changes with x:
1. Find how y changes with u: dy/du
2. Find how u changes with x: du/dx
3. Multiply them together: multiply rates of change
```

**Simple Numerical Example:**

```
y = (x + 2)┬▓

Method 1 (direct):
Expand: y = x┬▓ + 4x + 4
dy/dx = 2x + 4
At x=1: dy/dx = 6

Method 2 (chain rule):
Let u = x + 2 (inner function)
y = u┬▓ (outer function)

du/dx = 1 (how u changes with x)
dy/du = 2u = 2(x+2) = 2x + 4 (how y changes with u)

By chain rule:
dy/dx = (dy/du) ├Ч (du/dx)
      = (2x + 4) ├Ч 1
      = 2x + 4

At x=1: dy/dx = 6 тЬУ

Same answer! Chain rule works.
```

**Why Chain Rule Matters for Neural Networks:**

```
Neural network: composition of many functions!

Output = sigmoid(WтВГ ├Ч ReLU(WтВВ ├Ч ReLU(WтВБ ├Ч input + bтВБ) + bтВВ) + bтВГ)

Question: How does loss change with WтВБ?
Path: WтВБ тЖТ hidden1 тЖТ hidden2 тЖТ output тЖТ loss

Must trace through all layers!
Chain rule: multiply derivatives through each layer!

Without chain rule:
- Cannot efficiently compute тИВLoss/тИВWтВБ
- Would need numerical approximations (slow)
- Backpropagation impractical

With chain rule:
- Efficient computation
- Each layer contributes one multiplication
- Backpropagation practical and fast!
```

### 9.2 Why Chain Rule Is Required

**The Problem It Solves:**

```
Layer-wise functions compose:
aтВА = input
aтВБ = fтВБ(WтВБ ├Ч aтВА + bтВБ)
aтВВ = fтВВ(WтВВ ├Ч aтВБ + bтВВ)
aтВГ = fтВГ(WтВГ ├Ч aтВВ + bтВГ)
Loss = L(aтВГ, target)

To update WтВБ, need: тИВLoss/тИВWтВБ

Direct calculation extremely complex:
тИВLoss/тИВWтВБ involves all layers!
Loss depends on aтВГ
aтВГ depends on aтВВ
aтВВ depends on aтВБ
aтВБ depends on WтВБ

Chain rule breaks this into manageable pieces!
```

**Chain Rule Enables Efficient Computation:**

```
тИВLoss/тИВWтВБ = (тИВLoss/тИВaтВГ) ├Ч (тИВaтВГ/тИВaтВВ) ├Ч (тИВaтВВ/тИВaтВБ) ├Ч (тИВaтВБ/тИВWтВБ)

Computation:
1. тИВLoss/тИВaтВГ: how loss changes with final output (easy!)
2. тИВaтВГ/тИВaтВВ: how final layer output changes with its input (easy!)
3. тИВaтВВ/тИВaтВБ: how layer 2 output changes with its input (easy!)
4. тИВaтВБ/тИВWтВБ: how layer 1 output changes with WтВБ (easy!)
5. Multiply all together!

Each piece local and simple!
Composition gives global gradient!
This is backpropagation!
```

### 9.3 Derivative Calculations

**Common Derivatives (Activation Functions):**

```
Sigmoid: ╧Г(z) = 1/(1 + e^(-z))
Derivative: ╧Г'(z) = ╧Г(z) ├Ч (1 - ╧Г(z))

Tanh: tanh(z) = (e^z - e^(-z))/(e^z + e^(-z))
Derivative: tanh'(z) = 1 - tanh┬▓(z)

ReLU: ReLU(z) = max(0, z)
Derivative: ReLU'(z) = 1 if z > 0, else 0

Linear: f(z) = z
Derivative: f'(z) = 1
```

**Common Derivatives (Loss Functions):**

```
MSE: L = (y - ┼╖)┬▓
Derivative: тИВL/тИВ┼╖ = -2(y - ┼╖)

Binary Cross-Entropy: L = -[y├Чlog(┼╖) + (1-y)├Чlog(1-┼╖)]
Derivative: тИВL/тИВ┼╖ = (┼╖ - y) / (┼╖(1-┼╖))

Categorical Cross-Entropy: L = -╬г y_i ├Ч log(┼╖_i)
Derivative: тИВL/тИВ┼╖_i = -y_i / ┼╖_i

(Simplifies when using softmax!)
```

**Combining Derivatives (Chain Rule):**

```
For neuron:
z = W ├Ч x + b
a = f(z) = activation(z)

тИВa/тИВW = (тИВa/тИВz) ├Ч (тИВz/тИВW)
       = f'(z) ├Ч x
       
Meaning: 
- f'(z): how much activation responds to z
- x: how much z responds to W
- Product: total effect

Example with ReLU:
z = 2x + 3
a = ReLU(z) = max(0, 2x + 3)

If z > 0:
тИВa/тИВx = ReLU'(z) ├Ч 2 = 1 ├Ч 2 = 2

If z < 0:
тИВa/тИВx = ReLU'(z) ├Ч 2 = 0 ├Ч 2 = 0
```

### 9.4 Chain Rule Example with One Neuron

**Simple Example: Single Neuron Regression**

```
Setup:
- Input: x = 2
- Weight: w = 0.5
- Bias: b = 0.1
- Target: y = 5
- Activation: Linear (no activation)
- Loss: MSE

Forward Pass:
z = w ├Ч x + b = 0.5 ├Ч 2 + 0.1 = 1.1
a = z = 1.1 (linear activation)
┼╖ = 1.1 (prediction)
Loss = (y - ┼╖)┬▓ = (5 - 1.1)┬▓ = 15.21

Backward Pass (Computing тИВLoss/тИВw):

Step 1: тИВLoss/тИВ┼╖ (loss with respect to prediction)
Loss = (y - ┼╖)┬▓
тИВLoss/тИВ┼╖ = -2(y - ┼╖) = -2(5 - 1.1) = -2(3.9) = -7.8

Meaning: If increase prediction by 1, loss decreases by 7.8

Step 2: тИВ┼╖/тИВz (prediction with respect to z)
┼╖ = z (linear)
тИВ┼╖/тИВz = 1

Meaning: If increase z by 1, prediction increases by 1

Step 3: тИВz/тИВw (z with respect to weight)
z = w ├Ч x + b
тИВz/тИВw = x = 2

Meaning: If increase w by 1, z increases by 2 (because x=2)

Step 4: Chain Rule - Multiply All Together
тИВLoss/тИВw = (тИВLoss/тИВ┼╖) ├Ч (тИВ┼╖/тИВz) ├Ч (тИВz/тИВw)
         = (-7.8) ├Ч (1) ├Ч (2)
         = -15.6

Interpretation:
- Gradient is -15.6 (negative!)
- Negative means: increase w decreases loss
- Should increase w to reduce loss

Verification (numerical check):
Current: w = 0.5, Loss = 15.21
Try: w = 0.501 (slightly larger)
z = 0.501 ├Ч 2 + 0.1 = 1.102
┼╖ = 1.102
Loss = (5 - 1.102)┬▓ = 15.1904
Change in loss: 15.1904 - 15.21 = -0.0196 тЙИ -15.6 ├Ч 0.001 тЬУ

Gradient calculation correct!
```

### 9.5 Multi-Layer Chain Rule

**Example: Two Hidden Layers**

```
Architecture:
Input(1) тЖТ Hidden1(1) тЖТ Hidden2(1) тЖТ Output(1)

Parameters:
WтВБ = weight inputтЖТhidden1
bтВБ = bias hidden1
WтВВ = weight hidden1тЖТhidden2
bтВВ = bias hidden2
WтВГ = weight hidden2тЖТoutput
bтВГ = bias output

Forward Pass:
zтВБ = WтВБ ├Ч x + bтВБ
aтВБ = ReLU(zтВБ)
zтВВ = WтВВ ├Ч aтВБ + bтВВ
aтВВ = ReLU(zтВВ)
zтВГ = WтВГ ├Ч aтВВ + bтВГ
┼╖ = Sigmoid(zтВГ)
Loss = Binary_Cross_Entropy(y, ┼╖)

Backward Pass (for WтВВ as example):

тИВLoss/тИВWтВВ = (тИВLoss/тИВ┼╖) ├Ч (тИВ┼╖/тИВzтВГ) ├Ч (тИВzтВГ/тИВaтВВ) ├Ч (тИВaтВВ/тИВzтВВ) ├Ч (тИВzтВВ/тИВWтВВ)
            тФФтФАтФАтФАтФАтФАтФАтФАтФАтФАтФАтФмтФАтФАтФАтФАтФАтФАтФАтФАтФАтФАтФШ
              Gradient flows back through network
              Chain rule: multiply derivatives
```

**Key Insight:**

```
Each layer computes its local derivative:
- Output layer: тИВLoss/тИВ┼╖ and тИВ┼╖/тИВzтВГ
- Layer 3: тИВzтВГ/тИВaтВВ
- Layer 2: тИВaтВВ/тИВzтВВ and тИВzтВВ/тИВWтВВ
- Etc.

Global gradient: Product of all local gradients!
Backpropagation: compute these local gradients efficiently
Pass error signal backward through network
Each layer knows how much it contributed to error
```

## 10. Weight Update Process

### 10.1 Gradient Meaning

**What Does Gradient Tell Us?**

```
Gradient тИВLoss/тИВw = how loss changes when w changes

Numerically:
If тИВLoss/тИВw = 5:
- Increase w by 0.1 тЖТ Loss increases by ~0.5
- Increase w by -0.1 тЖТ Loss decreases by ~0.5
- Positive gradient: increasing w increases loss (bad)
- So: decrease w to reduce loss

If тИВLoss/тИВw = -3:
- Increase w by 0.1 тЖТ Loss decreases by ~0.3
- Increase w by -0.1 тЖТ Loss increases by ~0.3
- Negative gradient: increasing w decreases loss (good!)
- So: increase w to reduce loss

If тИВLoss/тИВw = 0:
- Small changes to w don't change loss (much)
- w is at critical point (minimum, maximum, or saddle)
- No clear direction to improve
```

**Magnitude Interpretation:**

```
Large gradient (e.g., 10):
- Loss very sensitive to w
- w needs significant change
- Currently far from optimal
- Take bigger steps

Small gradient (e.g., 0.1):
- Loss not very sensitive to w
- w nearly optimal
- Take smaller steps
- Fine-tuning stage

Zero gradient:
- w already optimal (or saddle point)
- No change needed

Gradient used as step size guidance!
Large gradient тЖТ large step
Small gradient тЖТ small step
```

### 10.2 Weight Update Formula

**The Update Rule:**

```
w_new = w_old - learning_rate ├Ч тИВLoss/тИВw

Where:
- w_old: current weight value
- w_new: new weight value
- learning_rate: hyperparameter (╬▒), typically 0.01 or smaller
- тИВLoss/тИВw: gradient (from backpropagation)

Key Ideas:
1. Subtract (not add): go opposite to gradient
   - Gradient shows direction of increase
   - We want to decrease loss
   - So subtract gradient

2. Multiply by learning_rate: control step size
   - learning_rate = 0.1, gradient = 2
   - Step = 0.1 ├Ч 2 = 0.2 (small controlled step)
   - Without learning_rate, step = 2 (huge, might overshoot)

3. Update all weights simultaneously
   - All weights updated using their own gradients
   - Single unified training step
```

**Intuition:**

```
Imagine descending a hill:
- Current position: w_old
- Slope at position: тИВLoss/тИВw (steepness)
- Opposite of slope: direction to go
- Step size: learning_rate

Strategy:
1. Look around (compute gradient)
2. Identify downhill direction (opposite to gradient)
3. Take a step (learning_rate controls step size)
4. Repeat

This is gradient descent!
```

### 10.3 Effect of Learning Rate

**Too Small Learning Rate:**

```
Example: learning_rate = 0.0001, gradient = -2

w_new = w_old - 0.0001 ├Ч (-2)
      = w_old + 0.0002

Very tiny change!

Epoch 1: w = 0 тЖТ 0.0002
Epoch 2: w = 0.0002 тЖТ 0.0004
Epoch 3: w = 0.0004 тЖТ 0.0006
...
Takes many epochs to reach optimal w!

Training very slow!
Might take 1,000,000 epochs to converge.

Loss Curve:
Loss
  |*
  | *
  |  *
  |   *
  |    *
  |____*__ Epoch
  Barely decreasing per epoch
  Takes forever!
```

**Too Large Learning Rate:**

```
Example: learning_rate = 1.0, gradient = -2

w_new = w_old - 1.0 ├Ч (-2)
      = w_old + 2.0

Huge jump!

Epoch 1: w = 0 тЖТ 2.0
Epoch 2: w = 2.0 тЖТ ?

If gradient at w=2 is positive (say +3):
w_new = 2.0 - 1.0 ├Ч 3 = -1.0

Oscillates! Jumps over minimum!

Epoch 3: w = -1.0 тЖТ ?
If gradient = -2:
w_new = -1.0 - 1.0 ├Ч (-2) = 1.0

Oscillates between 1, -1, 2, -1, ...
Never settles at optimum!

Loss Curve:
Loss
  |*
  | \* *
  |  \ * \*
  |   \  \ \*
  |____\__\____ Epoch
  Oscillates, diverges!
  Training unstable!
```

**Optimal Learning Rate:**

```
Example: learning_rate = 0.1, gradient = -2

w_new = w_old - 0.1 ├Ч (-2)
      = w_old + 0.2

Good step size!

Epoch 1: w = 0 тЖТ 0.2
Epoch 2: w = 0.2 тЖТ 0.4 (approximately, gradient smaller)
Epoch 3: w = 0.4 тЖТ 0.6
...
Converges smoothly!

Loss Curve:
Loss
  |*
  | *
  |  *
  |   *
  |    ***
  |_______*__ Epoch
  
  Smooth descent
  Steady convergence
  Reaches minimum in reasonable time
```

**Effect on Training:**

## Complete Practical Example: Two Hidden Layers

Now let's trace through a complete example with all three concepts together!

### Network Architecture:
```
Input (1 feature) тЖТ Hidden Layer 1 (2 neurons) тЖТ Hidden Layer 2 (1 neuron) тЖТ Output (1 neuron)

Parameters:
Input: x
Layer 1: WтВБ (2├Ч1), bтВБ (2├Ч1)
Layer 2: WтВВ (1├Ч2), bтВВ (1├Ч1)
Output: WтВГ (1├Ч1), bтВГ (1├Ч1)

Activation: ReLU for hidden, Linear for output
Loss: MSE (regression task)
```

### Initial Setup:

```
Training Data (single sample for simplicity):
x = 2.0 (input)
y = 10.0 (target)

Network Weights (initialized):
WтВБ = [0.5, -0.3]с╡А (shape 2├Ч1)
bтВБ = [0.1, -0.2]с╡А (shape 2├Ч1)
WтВВ = [0.4, 0.6] (shape 1├Ч2)
bтВВ = [0.05] (shape 1├Ч1)
WтВГ = [0.8] (shape 1├Ч1)
bтВГ = [0.02] (shape 1├Ч1)

Learning Rate: lr = 0.01
```

### Step 1: Forward Propagation

**Layer 1 Computation:**

```
zтВБ = WтВБ ├Ч x + bтВБ
   = [0.5, -0.3]с╡А ├Ч 2.0 + [0.1, -0.2]с╡А
   = [0.5├Ч2 + 0.1, -0.3├Ч2 - 0.2]с╡А
   = [1.0 + 0.1, -0.6 - 0.2]с╡А
   = [1.1, -0.8]с╡А

aтВБ = ReLU(zтВБ)
   = [ReLU(1.1), ReLU(-0.8)]с╡А
   = [1.1, 0]с╡А (negative value becomes 0)

Hidden Layer 1 Output: aтВБ = [1.1, 0]с╡А
```

**Layer 2 Computation:**

```
zтВВ = WтВВ ├Ч aтВБ + bтВВ
   = [0.4, 0.6] ├Ч [1.1, 0]с╡А + [0.05]
   = (0.4├Ч1.1 + 0.6├Ч0 + 0.05)
   = (0.44 + 0 + 0.05)
   = 0.49

aтВВ = ReLU(zтВВ)
   = ReLU(0.49)
   = 0.49 (positive, passes through)

Hidden Layer 2 Output: aтВВ = 0.49
```

**Output Layer:**

```
zтВГ = WтВГ ├Ч aтВВ + bтВГ
   = 0.8 ├Ч 0.49 + 0.02
   = 0.392 + 0.02
   = 0.412

┼╖ = zтВГ (linear activation for regression)
   = 0.412

Prediction: ┼╖ = 0.412
```

**Loss Calculation:**

```
Loss = (y - ┼╖)┬▓
     = (10.0 - 0.412)┬▓
     = (9.588)┬▓
     = 91.93

Current Loss: 91.93 (very high! Prediction far from target)
```

### Step 2: Backward Propagation

**Computing тИВLoss/тИВ┼╖ (Loss Gradient):**

```
Loss = (y - ┼╖)┬▓
тИВLoss/тИВ┼╖ = -2(y - ┼╖)
         = -2(10.0 - 0.412)
         = -2 ├Ч 9.588
         = -19.176

Interpretation: Increasing ┼╖ by 1 decreases loss by ~19.176
We should increase the prediction to reduce error!
```

**Computing тИВ┼╖/тИВzтВГ (Output Layer Activation):**

```
┼╖ = zтВГ (linear activation)
тИВ┼╖/тИВzтВГ = 1

(Linear activation: output same as input, derivative = 1)
```

**Computing тИВzтВГ/тИВWтВГ (Output Layer Weight):**

```
zтВГ = WтВГ ├Ч aтВВ + bтВГ
тИВzтВГ/тИВWтВГ = aтВВ = 0.49

Interpretation: If increase WтВГ by 1, zтВГ increases by aтВВ = 0.49
```

**Computing тИВzтВГ/тИВaтВВ (Output Layer Input):**

```
zтВГ = WтВГ ├Ч aтВВ + bтВГ
тИВzтВГ/тИВaтВВ = WтВГ = 0.8

Interpretation: If increase aтВВ by 1, zтВГ increases by WтВГ = 0.8
This error signal propagates back!
```

**Computing тИВaтВВ/тИВzтВВ (Hidden Layer 2 Activation):**

```
aтВВ = ReLU(zтВВ)
zтВВ = 0.49 > 0, so ReLU passes through
тИВaтВВ/тИВzтВВ = 1

(ReLU derivative: 1 for positive input)
```

**Computing тИВzтВВ/тИВWтВВ (Hidden Layer 2 Weight):**

```
zтВВ = WтВВ ├Ч aтВБ + bтВВ
For each element of WтВВ:
тИВzтВВ/тИВWтВВ[0] = aтВБ[0] = 1.1
тИВzтВВ/тИВWтВВ[1] = aтВБ[1] = 0

WтВВ = [0.4, 0.6]
тИВzтВВ/тИВWтВВ = [1.1, 0]с╡А (same as aтВБ)
```

**Computing тИВzтВВ/тИВaтВБ (Hidden Layer 2 Input):**

```
zтВВ = WтВВ ├Ч aтВБ + bтВВ
тИВzтВВ/тИВaтВБ = WтВВс╡А = [0.4, 0.6]

This is the error signal passed back to hidden layer 1!
```

**Computing тИВaтВБ/тИВzтВБ (Hidden Layer 1 Activation):**

```
aтВБ = ReLU(zтВБ)
zтВБ = [1.1, -0.8]с╡А

For element 1: zтВБ[0] = 1.1 > 0, so тИВaтВБ[0]/тИВzтВБ[0] = 1
For element 2: zтВБ[1] = -0.8 < 0, so тИВaтВБ[1]/тИВzтВБ[1] = 0

тИВaтВБ/тИВzтВБ = [1, 0]с╡А (ReLU kills negative signals!)
```

**Computing тИВzтВБ/тИВWтВБ (Hidden Layer 1 Weight):**

```
zтВБ = WтВБ ├Ч x + bтВБ
For each element of WтВБ:
тИВzтВБ/тИВWтВБ[0] = x = 2.0
тИВzтВБ/тИВWтВБ[1] = x = 2.0

тИВzтВБ/тИВWтВБ = [2.0, 2.0]с╡А
```

### Step 3: Applying Chain Rule - Computing Gradients for Each Weight

**Gradient for WтВГ:**

```
тИВLoss/тИВWтВГ = (тИВLoss/тИВ┼╖) ├Ч (тИВ┼╖/тИВzтВГ) ├Ч (тИВzтВГ/тИВWтВГ)
          = (-19.176) ├Ч (1) ├Ч (0.49)
          = -9.396

Chain Rule: Multiply derivatives through each layer!
```

**Gradient for WтВВ[0]:**

```
тИВLoss/тИВWтВВ[0] = (тИВLoss/тИВ┼╖) ├Ч (тИВ┼╖/тИВzтВГ) ├Ч (тИВzтВГ/тИВaтВВ) ├Ч (тИВaтВВ/тИВzтВВ) ├Ч (тИВzтВВ/тИВWтВВ[0])
             = (-19.176) ├Ч (1) ├Ч (0.8) ├Ч (1) ├Ч (1.1)
             = -19.176 ├Ч 0.8 ├Ч 1.1
             = -16.875

Note: Error signal: -19.176 flows back
      Multiplied by WтВГ = 0.8: becomes -15.341
      Multiplied by ReLU derivative = 1: stays -15.341
      Multiplied by aтВБ[0] = 1.1: becomes -16.875
```

**Gradient for WтВВ[1]:**

```
тИВLoss/тИВWтВВ[1] = (тИВLoss/тИВ┼╖) ├Ч (тИВ┼╖/тИВzтВГ) ├Ч (тИВzтВГ/тИВaтВВ) ├Ч (тИВaтВВ/тИВzтВВ) ├Ч (тИВzтВВ/тИВWтВВ[1])
             = (-19.176) ├Ч (1) ├Ч (0.8) ├Ч (1) ├Ч (0)
             = 0

Note: aтВБ[1] = 0 (neuron was inactive in layer 1)
      Zero activation ├Ч gradient = zero contribution
      WтВВ[1] receives no gradient signal!
```

**Gradient for WтВБ[0]:**

```
тИВLoss/тИВWтВБ[0] = (тИВLoss/тИВ┼╖) ├Ч (тИВ┼╖/тИВzтВГ) ├Ч (тИВzтВГ/тИВaтВВ) ├Ч (тИВaтВВ/тИВzтВВ) ├Ч (тИВzтВВ/тИВaтВБ[0]) ├Ч (тИВaтВБ[0]/тИВzтВБ[0]) ├Ч (тИВzтВБ[0]/тИВWтВБ[0])
             = (-19.176) ├Ч (1) ├Ч (0.8) ├Ч (1) ├Ч (0.4) ├Ч (1) ├Ч (2.0)
             = -12.272

Note: Error flows all the way back: -19.176
      Through WтВГ: ├Ч 0.8 = -15.341
      Through WтВВ[0]: ├Ч 0.4 = -6.136
      Through ReLU (active): ├Ч 1 = -6.136
      Through input x: ├Ч 2.0 = -12.272

Each layer multiplies error signal!
```

**Gradient for WтВБ[1]:**

```
тИВLoss/тИВWтВБ[1] = (тИВLoss/тИВ┼╖) ├Ч (тИВ┼╖/тИВzтВГ) ├Ч (тИВzтВГ/тИВaтВВ) ├Ч (тИВaтВВ/тИВzтВВ) ├Ч (тИВzтВВ/тИВaтВБ[1]) ├Ч (тИВaтВБ[1]/тИВzтВБ[1]) ├Ч (тИВzтВБ[1]/тИВWтВБ[1])
             = (-19.176) ├Ч (1) ├Ч (0.8) ├Ч (1) ├Ч (0.6) ├Ч (0) ├Ч (2.0)
             = 0

Note: aтВБ[1] = 0 (was negative, killed by ReLU)
      тИВaтВБ[1]/тИВzтВБ[1] = 0 (ReLU blocked gradient)
      Zero in chain = entire gradient becomes 0!

This is vanishing gradient for dead neurons!
WтВБ[1] cannot learn when neuron is inactive!
```

### Summary of Gradients:

```
тИВLoss/тИВWтВГ = -9.396
тИВLoss/тИВWтВВ[0] = -16.875
тИВLoss/тИВWтВВ[1] = 0 (zero because aтВБ[1] = 0)
тИВLoss/тИВWтВБ[0] = -12.272
тИВLoss/тИВWтВБ[1] = 0 (zero because neuron 1 was dead)

Biases (similar process):
тИВLoss/тИВbтВГ = -19.176
тИВLoss/тИВbтВВ = -15.341
тИВLoss/тИВbтВБ[0] = -6.136
тИВLoss/тИВbтВБ[1] = 0 (zero because neuron 1 was dead)
```

### Step 4: Weight Update

**Update WтВГ:**

```
wтВГ_new = wтВГ_old - learning_rate ├Ч тИВLoss/тИВwтВГ
       = 0.8 - 0.01 ├Ч (-9.396)
       = 0.8 + 0.09396
       = 0.894

Change: +0.094 (positive because gradient negative!)
Effect: Increased WтВГ (which increases output)
Consequence: Next prediction will be larger, closer to target!
```

**Update WтВВ:**

```
wтВВтВА_new = wтВВтВА_old - learning_rate ├Ч тИВLoss/тИВwтВВтВА
        = 0.4 - 0.01 ├Ч (-16.875)
        = 0.4 + 0.169
        = 0.569

wтВВтВБ_new = wтВВтВБ_old - learning_rate ├Ч тИВLoss/тИВwтВВтВБ
        = 0.6 - 0.01 ├Ч (0)
        = 0.6

Note: WтВВ[1] doesn't update (gradient = 0)
      Because its input (aтВБ[1]) was 0
      Nothing learned about this weight this iteration!
```

**Update WтВБ:**

```
wтВБтВА_new = wтВБтВА_old - learning_rate ├Ч тИВLoss/тИВwтВБтВА
        = 0.5 - 0.01 ├Ч (-12.272)
        = 0.5 + 0.123
        = 0.623

wтВБтВБ_new = wтВБтВБ_old - learning_rate ├Ч тИВLoss/тИВwтВБтВБ
        = -0.3 - 0.01 ├Ч (0)
        = -0.3

Note: WтВБ[1] doesn't update (gradient = 0)
      Because ReLU killed it (output was 0)
      This neuron can't learn!

This is the dead neuron problem!
```

**Update Biases:**

```
bтВГ_new = 0.02 - 0.01 ├Ч (-19.176) = 0.02 + 0.192 = 0.212
bтВВ_new = 0.05 - 0.01 ├Ч (-15.341) = 0.05 + 0.153 = 0.203
bтВБ[0]_new = 0.1 - 0.01 ├Ч (-6.136) = 0.1 + 0.061 = 0.161
bтВБ[1]_new = -0.2 - 0.01 ├Ч (0) = -0.2 (no update)
```

### Updated Network After One Iteration:

```
WтВБ_new = [0.623, -0.3]с╡А
bтВБ_new = [0.161, -0.2]с╡А
WтВВ_new = [0.569, 0.6]
bтВВ_new = [0.203]
WтВГ_new = [0.894]
bтВГ_new = [0.212]

All weights changed to reduce loss!
Some weights (those in dead neurons) didn't change.
```

### Step 5: New Forward Pass with Updated Weights

```
zтВБ_new = [0.623, -0.3]с╡А ├Ч 2.0 + [0.161, -0.2]с╡А
       = [1.246 + 0.161, -0.6 - 0.2]с╡А
       = [1.407, -0.8]с╡А

aтВБ_new = ReLU([1.407, -0.8]с╡А)
       = [1.407, 0]с╡А (second neuron still dead!)

zтВВ_new = [0.569, 0.6] ├Ч [1.407, 0]с╡А + [0.203]
       = 0.801 + 0.203
       = 1.004

aтВВ_new = ReLU(1.004) = 1.004

zтВГ_new = 0.894 ├Ч 1.004 + 0.212
       = 0.898 + 0.212
       = 1.110

┼╖_new = 1.110

Loss_new = (10.0 - 1.110)┬▓
         = (8.89)┬▓
         = 79.03

Previous Loss: 91.93
New Loss: 79.03
Improvement: 91.93 - 79.03 = 12.90 тЬУ

Loss decreased! Weights moved in right direction!
```

### Observations from This Example:

```
1. Backpropagation computes gradients for all weights
   тЬУ Error signal flows backward through network
   тЬУ Each layer computes local gradients
   тЬУ Chain rule multiplies them together

2. Chain rule enables efficient computation
   тЬУ Don't need to differentiate entire network at once
   тЬУ Each layer contributes one multiplication
   тЬУ O(n) complexity, not O(n┬▓)

3. Gradient descent uses these gradients to update
   тЬУ w_new = w_old - lr ├Ч тИВLoss/тИВw
   тЬУ Negative gradients тЖТ increase weight
   тЬУ Positive gradients тЖТ decrease weight
   тЬУ Magnitude of gradient тЖТ size of update

4. Dead neurons prevent learning
   тЬУ Neuron 2 in layer 1: output = 0
   тЬУ Gradient through it: 0
   тЬУ Weights connected to it don't update
   тЬУ Can't escape dead neuron state!
   
   This is why Leaky ReLU better:
   тЬУ Even negative inputs produce small signal
   тЬУ Gradient never exactly zero
   тЬУ Neurons can recover

5. Learning rate controls step size
   тЬУ WтВГ: gradient -9.396 тЖТ step 0.094
   тЬУ Learning rate 0.01 makes step reasonable
   тЬУ Too high: might overshoot minimum
   тЬУ Too low: convergence very slow

6. Multiple iterations needed
   тЬУ After 1 iteration: Loss dropped 91.93 тЖТ 79.03
   тЬУ Still far from target (10.0)
   тЬУ Need hundreds/thousands more iterations
   тЬУ Each iteration: loss decreases slightly
   тЬУ Eventually converges to minimum
```

### Mathematical Summary:

```
Backpropagation Formula (for this example):

тИВLoss/тИВw = (тИВLoss/тИВ┼╖) ├Ч (тИВ┼╖/тИВzтВГ) ├Ч (тИВzтВГ/тИВaтВВ) ├Ч (тИВaтВВ/тИВzтВВ) ├Ч ... ├Ч (тИВz_i/тИВw)

General Pattern:
- Start at loss (last layer)
- Work backward through network
- Multiply local derivatives
- Reach desired weight

Each layer computes:
тИВa/тИВz = f'(z) [activation derivative]
тИВz/тИВw = input [pre-synaptic activation]

Product of all: total gradient!

Weight update:
w тЖР w - learning_rate ├Ч gradient

This simple rule, repeated millions of times
Trains entire network!
```

### Complete Training Process (Multiple Iterations):

```
Iteration 1:
  Loss: 91.93 тЖТ 79.03 (improvement: -12.90)

Iteration 2:
  Forward pass with updated weights
  Compute new loss
  Backprop: compute new gradients
  Update weights again
  Loss: 79.03 тЖТ 68.5 (improvement: -10.53)

Iteration 3:
  Loss: 68.5 тЖТ 59.2 (improvement: -9.3)

...

Iteration 100:
  Loss: ~0.5 (very close!)

Iteration 500:
  Loss: ~0.001 (nearly converged!)

Each iteration:
- Forward pass: compute prediction
- Loss: measure error
- Backward pass: compute gradients
- Weight update: move in right direction
- Loss decreases (overall trend)
- Eventually converges

This is how neural networks learn!
```
