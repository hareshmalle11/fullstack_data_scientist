### 🔤 Tokenization Methods in NLTK

### 1) `word_tokenize()`
- Splits text into individual word tokens.
- Separates punctuation into separate tokens.
- Uses **Punkt + Treebank** internally.
- Commonly used for general NLP preprocessing.

---

### 2) `sent_tokenize()`
- Splits text into complete sentences.
- Detects sentence boundaries such as `.`, `?`, `!`.
- Uses the **Punkt sentence tokenizer** model.
- Useful for summarization and document segmentation.

---

### 3) `WordPunctTokenizer`
- Splits text using whitespace and punctuation rules.
- More aggressive than `word_tokenize()`.
- Breaks words like `can't` into `can`, `'`, `t`.
- Useful for fast, rule-based tokenization.

---

### 4) `TreebankWordTokenizer`
- Based on **Penn Treebank** tokenization rules.
- Handles contractions linguistically (e.g., `can't` → `ca`, `n't`).
- Better suited for POS tagging and parsing tasks.
- More linguistically accurate than `WordPunctTokenizer`.



# Stemming 
- it is a processing of word in to its word stem
- stemming is important in NLU and NLp
- ex: eating,eat,eaten------->eat
- ex: going,gone,goes-------->go


```python
#import stem
from nltk.stem import PorterStemmer
ps = PorterStemmer()
# stemming
words = ['eaten', 'eats', 'eaten', 'writer', 'programming', 
         'programmer', 'history', 'finally', 'finalized']
for word in words:
    print(word+"-------->"+ps.stem(word))
```

# Regular Expressions

```python
from nltk.stem import RegexpStemmer
reg=RegexpStemmer('ing$|s$|s$|able$|',min=4)
# regex stemmer
print(reg.stem('eating'))
print(reg.stem('programming'))
```

```python
from nltk.stem import SnowballStemmer
# snowball stemmer
sno = SnowballStemmer('english')
for word in words:
    print(word+"-------->"+sno.stem(word))
```

# What is Lemmatization?
- Lemmatization is the process of converting a word into its correct base form (dictionary form).
- The base form is called a lemma. 

| Stemming            | Lemmatization      |
| ------------------- | ------------------ |
| Cuts word blindly   | Uses dictionary    |
| May give wrong word | Gives correct word |
| Faster              | Slightly slower    |

# TF-IDF(Term Frequency-Inverse Document Frequency)

### Term Frequency (TF)

**TF = (Number of times the word appears in the sentence) / (Total number of words in the sentence)**

---

### Inverse Document Frequency (IDF)

**IDF = (Total number of sentences) / (Number of sentences containing the word)**


## Example: TF and IDF Calculation

### Sentences
- S1: good boy  
- S2: good girl  
- S3: good boy girl  

Total sentences = 3

---

## Term Frequency (TF)

| Word | S1 | S2 | S3 |
|-----|----|----|----|
| good | 1/2 | 1/2 | 1/3 |
| boy  | 1/2 | 0   | 1/3 |
| girl | 0   | 1/2 | 1/3 |

Explanation:
- S1 has 2 words → good, boy  
- S2 has 2 words → good, girl  
- S3 has 3 words → good, boy, girl  

TF = (count of word in sentence) / (total words in sentence)

---

## Inverse Document Frequency (IDF)

| Word | Sentences Containing Word | IDF |
|-----|---------------------------|-----|
| good | 3 | log₂(3/3) = 0 |
| boy  | 2 | log₂(3/2) |
| girl | 2 | log₂(3/2) |

IDF Formula:

IDF(word) = log₂ ( Total Sentences / Sentences Containing Word )

## CBOW (Continuous Bag of Words)

### Example Sentence
Sentence used:

inventory company is related to data science

Words:

[inventory, company, is, related, to, data, science]

### Window Size = 2
Window size means how many surrounding words are used as context.

If window size = 2:
- 2 words before the target word
- 2 words after the target word

### Example 1
Target word (Output):

is

Context words (Input):

[inventory, company, related, to]

Training pair:

Input  → [inventory, company, related, to]  
Output → is

The model predicts the center word using surrounding words.

### Example 2 (Sliding Window)

Target word:

related

Context words:

[company, is, to, data]

Training pair:

Input  → [company, is, to, data]  
Output → related

### Example 3

Target word:

to

Context words:

[is, related, data, science]

Training pair:

Input  → [is, related, data, science]  
Output → to

### Key Idea

CBOW uses surrounding words (context) to predict the target word.

Input  → Context words  
Output → Target word
