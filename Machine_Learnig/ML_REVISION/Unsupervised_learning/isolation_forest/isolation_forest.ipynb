# Isolation Forest Anomaly Detection on Customer Credit Card Data
This notebook preprocesses the data, fits an Isolation Forest model to detect outliers, and saves the trained pipeline.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import IsolationForest
import joblib
```

```python
df = pd.read_csv('CC GENERAL.csv')
X = df.drop(columns=['CUST_ID'])

preprocessor = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
X_scaled = preprocessor.fit_transform(X)
```

```python
# Fit Isolation Forest (contamination=0.05 representing 5% expected outliers)
iso_forest = IsolationForest(contamination=0.05, random_state=42)
labels = iso_forest.fit_predict(X_scaled)
print('Outliers detected:', np.sum(labels == -1))
```

```python
# Save pipeline
iso_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', IsolationForest(contamination=0.05, random_state=42))
])
iso_pipeline.fit(X)
joblib.dump(iso_pipeline, 'isolation_forest_model.pkl')
print('Isolation Forest Pipeline saved successfully!')
```
