# t-SNE Dimensionality Reduction on Customer Credit Card Data
This notebook preprocesses the data, runs t-SNE to construct a 2D visual projection of the customer profiles, and saves the final pre-calculated dataset.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.manifold import TSNE
import joblib
```

```python
df = pd.read_csv('CC GENERAL.csv')
X = df.drop(columns=['CUST_ID'])

preprocessor = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
X_scaled = preprocessor.fit_transform(X)
```

```python
# Run t-SNE (limit iterations/perplexity to default configuration)
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
tsne_results = tsne.fit_transform(X_scaled)
```

```python
# Save the precomputed dataset along with t-SNE coordinates for interactive app exploration
tsne_df = df.copy()
tsne_df['TSNE1'] = tsne_results[:, 0]
tsne_df['TSNE2'] = tsne_results[:, 1]
tsne_df['MINIMUM_PAYMENTS'] = tsne_df['MINIMUM_PAYMENTS'].fillna(tsne_df['MINIMUM_PAYMENTS'].median())
tsne_df['CREDIT_LIMIT'] = tsne_df['CREDIT_LIMIT'].fillna(tsne_df['CREDIT_LIMIT'].median())
joblib.dump(tsne_df, 'tsne_data.pkl')
print('t-SNE precomputed coordinate dataset saved successfully!')
```
