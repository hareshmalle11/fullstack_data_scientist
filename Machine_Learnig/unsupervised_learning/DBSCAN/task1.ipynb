```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.ensemble import IsolationForest
```

```python
df=pd.read_csv('task1.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# check for missing values
print(df.isnull().sum())
```

```python
# turn boolen values into integers
for col in df.columns:
    if df[col].dtype == 'bool':
        df[col] = df[col].astype(int)
        

```

```python
from sklearn.preprocessing import StandardScaler
# standardize the features
scaler = StandardScaler()
fea = scaler.fit_transform(df)
features = pd.DataFrame(fea, columns=df.columns)

```

```python
df=features
```

```python
df.info()
```

```python
# parameters for isolation forest
n_estimators = 100   # no:of trees in the forest
contamination = 0.04  # expected proportion of outliers in the data
sample_size = 256   # number of samples to draw from the data to train each base estimator
```

```python
# train the isolation forest model
model = IsolationForest(n_estimators=n_estimators, contamination=contamination, max_samples=sample_size, random_state=42)
model.fit(df)


# predict anomalies
anomaly_labels = model.predict(df)

# -1 indicates an anomaly, 1 indicates normal stored the anomaly labels in the original dataframe
df['anomaly'] = anomaly_labels
```

```python
df.head()
```

```python
# count ananomalies
anomaly_counts = df['anomaly'].value_counts()
print(anomaly_counts)
```

```python
#scatter plot (use existing feature names)
plt.figure(figsize=(10, 6))
normal = df[df['anomaly'] == 1]
anomalies = df[df['anomaly'] == -1]
plt.scatter(normal['packet_size'], normal['inter_arrival_time'], color='blue', label='Normal', alpha=0.5)
plt.scatter(anomalies['packet_size'], anomalies['inter_arrival_time'], color='red', label='Anomaly', alpha=0.5)
plt.xlabel('Packet Size')
plt.ylabel('Inter Arrival Time')
plt.title('Anomaly Detection using Isolation Forest')
plt.legend()
plt.show()
```
