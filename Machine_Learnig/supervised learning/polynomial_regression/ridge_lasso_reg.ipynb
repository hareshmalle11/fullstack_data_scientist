```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
dataset = pd.read_csv("CarPrice_Assignment.csv")
numerical_features = dataset.select_dtypes(include=[np.number])
```

```python
from sklearn.linear_model import Ridge
x2=numerical_features.drop(columns=['price'])
y2 = numerical_features['price']
# print(x2.head())
# print(y2.head())
x2_train, x2_test, y2_train, y2_test = train_test_split(
    x2, y2, test_size=0.2, random_state=42
)
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(x2_train, y2_train)
ridge_predictions = ridge_model.predict(x2_test)
ridge_r2 = r2_score(y2_test, ridge_predictions)
print(f"Ridge Regression R2 Score on Test Set: {ridge_r2}")
```

```python
#lasso regression
from sklearn.linear_model import Lasso
x3=numerical_features.drop(columns=['price'])
y3 = numerical_features['price']
x3_train, x3_test, y3_train, y3_test = train_test_split(
    x3, y3, test_size=0.2, random_state=42
)
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(x3_train, y3_train)
lasso_predictions = lasso_model.predict(x3_test)
lasso_r2 = r2_score(y3_test, lasso_predictions)
print(f"Lasso Regression R2 Score on Test Set: {lasso_r2}")
```

```python
from sklearn.linear_model import RidgeCV
alphas=np.arange(0.1,10,0.05)
scores=[]
for i in alphas:
    ridge_cv_model = RidgeCV(alphas=[i])
    ridge_cv_model.fit(x2_train, y2_train)
    ridge_cv_predictions = ridge_cv_model.predict(x2_test)
    ridge_cv_r2 = r2_score(y2_test, ridge_cv_predictions)
    scores.append((i, ridge_cv_r2))
    #print(f"RidgeCV Regression R2 Score on Test Set with alpha {i}: {ridge_cv_r2}")
for i,j in scores:
    if j==max([score[1] for score in scores]):
        print(f"Best alpha: {i:.2f} with R2 Score: {j}")
```

```python
from sklearn.linear_model import LassoCV
alphas=np.arange(0.1,10,0.05)
scores_lasso=[]
for i in alphas:
    lasso_cv_model = LassoCV(alphas=[i], max_iter=10000)
    lasso_cv_model.fit(x3_train, y3_train)
    lasso_cv_predictions = lasso_cv_model.predict(x3_test)
    lasso_cv_r2 = r2_score(y3_test, lasso_cv_predictions)
    scores_lasso.append((i, lasso_cv_r2))
    #print(f"LassoCV Regression R2 Score on Test Set with alpha {i}: {lasso_cv_r2}")
for i,j in scores_lasso:
    if j==max([score[1] for score in scores_lasso]):
        print(f"Best alpha: {i:.4f} with R2 Score: {j}")

```
