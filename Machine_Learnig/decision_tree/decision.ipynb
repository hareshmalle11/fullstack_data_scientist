#  Introduction to Decision Trees (Machine Learning)

## What is a Decision Tree?
A **Decision Tree** is a **supervised machine learning algorithm** used for:
- **Classification** (predicting categories)
- **Regression** (predicting continuous values)

It works like a **tree-shaped flowchart**, where decisions are made step by step based on feature values.

---

## How Does a Decision Tree Work?
A decision tree:
1. Starts with the **entire dataset** at the root node
2. Splits the data based on a **feature condition**
3. Repeats splitting to create **branches**
4. Ends at **leaf nodes**, which give the final prediction

The goal is to split data in a way that makes the output as **pure and accurate** as possible.

---

## Structure of a Decision Tree
- **Root Node**  The top node representing the full dataset
- **Decision Node**  A node that applies a condition on a feature
- **Branch**  Outcome of a decision
- **Leaf Node**  Final prediction (class or value)

---

## Types of Decision Trees

### 1. Classification Tree
- Output is **categorical**
- Examples:
  - Spam / Not Spam
  - Pass / Fail
  - Yes / No

### 2. Regression Tree
- Output is **numerical**
- Examples:
  - House price
  - Salary prediction

---

## Splitting Criteria
To decide the best split, decision trees use measures like:
- **Entropy**
- **Information Gain**
- **Gini Index**
- **Mean Squared Error (for regression)**

The feature that produces the **best split** is chosen at each step.

---

## Stopping Conditions
Tree growth stops when:
- All data belongs to one class
- Maximum depth is reached
- Minimum samples per node condition is met
- No further improvement is possible

---

## Advantages of Decision Trees
- Easy to understand and interpret
- No need for feature scaling
- Handles both numerical and categorical data
- Works well with non-linear data

---

## Disadvantages of Decision Trees
- Can easily **overfit**
- Sensitive to small changes in data
- Less accurate compared to ensemble models

---

## Real-World Applications
- Medical diagnosis
- Credit approval systems
- Customer churn prediction
- Recommendation systems

---

## Conclusion
Decision Trees are simple yet powerful models that mimic **human decision-making**.  
They form the foundation for advanced algorithms like **Random Forest** and **Boosting models**.


#  Splitting Criteria in Decision Trees

Decision Trees use **splitting criteria** to decide **which feature** should be used to split the data at each node.  
The goal is to create **purer child nodes** than the parent node.

The most commonly used measures are:
- **Entropy**
- **Information Gain**
- **Gini Index**

---

##  Entropy

### What is Entropy?
**Entropy** measures the **impurity or randomness** in a dataset.

- Low entropy  data is mostly one class (pure)
- High entropy  data is mixed (impure)

### Entropy Formula
\[
Entropy(S) = - \sum_{i=1}^{n} p_i \log_2(p_i)
\]

Where:
- \(p_i\) = proportion of class *i* in dataset *S*
- \(n\) = number of classes

---

### Entropy Values
| Situation | Entropy |
|--------|---------|
| All samples belong to one class | 0 |
| Equal samples in all classes | Maximum |

---

### Example of Entropy
Suppose a dataset has:
- 8 Yes
- 2 No

\[
Entropy = -(0.8\log_2 0.8 + 0.2\log_2 0.2) \approx 0.72
\]

---

##  Information Gain

### What is Information Gain?
**Information Gain (IG)** tells us **how much entropy is reduced** after splitting the dataset using a feature.

 The feature with **maximum Information Gain** is chosen.

---

### Information Gain Formula
\[
IG(S, A) = Entropy(S) - \sum_{v \in A} \frac{|S_v|}{|S|} Entropy(S_v)
\]

Where:
- \(S\) = original dataset
- \(A\) = feature used for splitting
- \(S_v\) = subset after split

---

### Intuition
- High IG  better split
- Low IG  weak split

---

### Example (Simple)
If:
- Parent entropy = 0.94
- Child weighted entropy = 0.40

\[
IG = 0.94 - 0.40 = 0.54
\]

So this feature is a **good splitting choice**.

---

##  Gini Index

### What is Gini Index?
**Gini Index** measures how often a randomly chosen element would be **incorrectly classified**.

- Lower Gini  better split
- Higher Gini  more impurity

---

### Gini Index Formula
\[
Gini(S) = 1 - \sum_{i=1}^{n} p_i^2
\]

Where:
- \(p_i\) = probability of class *i*

---

### Gini Values
| Situation | Gini |
|--------|------|
| All samples in one class | 0 |
| Equal class distribution | Maximum |

---

### Example of Gini Index
For:
- 70% Class A
- 30% Class B

\[
Gini = 1 - (0.7^2 + 0.3^2) = 0.42
\]

---

##  Entropy vs Gini Index

| Feature | Entropy | Gini Index |
|------|--------|------------|
| Measure | Disorder | Misclassification |
| Formula Complexity | More complex | Simpler |
| Range | 0 to 1 | 0 to 0.5 (binary) |
| Used in | ID3, C4.5 | CART |
| Speed | Slower | Faster |

---

##  Key Takeaways
- **Entropy** measures impurity
- **Information Gain** selects the best feature
- **Gini Index** is a faster alternative to entropy
- Decision Trees aim to **maximize IG** or **minimize Gini**

---

##  Conclusion
Entropy, Information Gain, and Gini Index are fundamental concepts that help Decision Trees decide **how and where to split data** to make accurate predictions.
