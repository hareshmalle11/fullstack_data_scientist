```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0,random_state=42)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVC
model=SVC(kernel= 'linear' )
model.fit(x_train,y_train)
model.coef_
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
#same with rbf kernal
model_rbf=SVC(kernel= 'rbf' )
model_rbf.fit(x_train,y_train)
y_pred_rbf=model_rbf.predict(x_test)
print(accuracy_score(y_test,y_pred_rbf))
print(classification_report(y_test,y_pred_rbf))
```

```python
#kernal sigmoid
model_sigmoid=SVC(kernel= 'sigmoid' )
model_sigmoid.fit(x_train,y_train)
y_pred_sigmoid=model_sigmoid.predict(x_test)
print(accuracy_score(y_test,y_pred_sigmoid))
print(classification_report(y_test,y_pred_sigmoid))
```

```python
#kernal poolynomial
model_poly=SVC(kernel= 'poly' )
model_poly.fit(x_train,y_train)
y_pred_poly=model_poly.predict(x_test)
print(accuracy_score(y_test,y_pred_poly))
print(classification_report(y_test,y_pred_poly))
```

```python
#hyperparameter tuning
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['rbf']}
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
grid.fit(x_train,y_train)
grid.best_params_
```
