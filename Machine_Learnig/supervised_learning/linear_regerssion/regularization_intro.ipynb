## 1. Introduction to Regularization

### What is Regularization?

**Regularization** is a technique used to prevent **overfitting** by adding a penalty term to the loss function. It discourages the model from learning overly complex patterns that may not generalize well to new data.

### The Core Idea

Instead of just minimizing the error, we minimize:

```
Error + Penalty for Complexity
```

### Why Do We Need It?

**Problems with standard linear regression:**
1. **Overfitting**: Model fits training data too well but performs poorly on new data
2. **High variance**: Small changes in data cause large changes in model
3. **Large coefficients**: Unstable predictions
4. **Multicollinearity**: When features are correlated, coefficients become unreliable

**Regularization solves these by:**
- Shrinking coefficient values
- Creating simpler, more generalizable models
- Reducing model variance
- Handling correlated features better

## 2. The Overfitting Problem

### Understanding Overfitting

**Underfitting** ← **Good Fit** → **Overfitting**

```
Model Complexity:   Low          Medium        High
Training Error:     High         Low           Very Low
Test Error:         High         Low           High
```

### Visual Example

Imagine fitting a polynomial to data:

**Degree 1 (Underfitting):**
```
     y
     |     •
     |  •     •
     | /   •
     |/  •
     +------------- x
     Simple line - misses pattern
```

**Degree 2 (Good Fit):**
```
     y
     |     •
     |  • ⌢ •
     | ⌣   •
     |  •
     +------------- x
     Captures trend well
```

**Degree 10 (Overfitting):**
```
     y
     |     •
     |  •⌢⌣⌢•
     | ⌣⌢ ⌣•
     |  •
     +------------- x
     Fits every point - unstable
```

### Mathematical Example

**Dataset:** 5 points with noise

Without regularization, fitting a high-degree polynomial:
```
ŷ = 100 + 0.5x - 0.001x² + 0.00003x³ - 0.0000004x⁴ + ...
```
**Problems:**
- Huge coefficients
- Perfect fit on training data (R² = 1.0)
- Poor performance on test data
- Sensitive to small data changes

With regularization:
```
ŷ = 95 + 0.3x - 0.0001x² + 0.000001x³
```
**Benefits:**
- Smaller coefficients
- Good fit on training data (R² = 0.95)
- Better performance on test data
- More stable predictions

## 3. How Regularization Works

### Standard Linear Regression (OLS)

**Objective:** Minimize Sum of Squared Errors (SSE)

```
Minimize: SSE = Σ(yᵢ - ŷᵢ)²
```

**Problem:** No constraint on coefficient size.

### Regularized Regression

**Objective:** Minimize SSE + Penalty

```
Minimize: SSE + λ × Penalty(coefficients)
```

Where:
- **λ (lambda)**: Regularization parameter (hyperparameter)
  - λ = 0: No regularization (standard OLS)
  - λ → ∞: Maximum regularization (all coefficients → 0)
  - 0 < λ < ∞: Balance between fit and simplicity

### The Trade-off

```
Total Cost = Goodness of Fit + Complexity Penalty
           = Σ(yᵢ - ŷᵢ)²  +  λ × f(β)
           
           ↑                  ↑
     Want to minimize    Want to minimize
     (better predictions) (simpler model)
```

**Key Insight:** Regularization biases the model to prefer smaller coefficients, trading some training accuracy for better generalization.

### Bias-Variance Tradeoff

**Without Regularization (λ = 0):**
- Low bias (fits training data well)
- High variance (sensitive to data changes)
- Risk of overfitting

**With Strong Regularization (λ large):**
- High bias (underfits training data)
- Low variance (stable predictions)
- Risk of underfitting

**Optimal Regularization (λ moderate):**
- Balanced bias and variance
- Good generalization
- Best test performance

```
Error
  |                    
  |         \         / Total Error
  |          \       /
  |           \     /
  |            \   /  ← Optimal λ
  |             \ /
  |         Bias²  \
  |                 \
  |                  Variance
  +----------------------------
       0        λ*        ∞
            λ (regularization)
```

## 4. Ridge Regression (L2 Regularization)

### Definition

Ridge regression adds an **L2 penalty** (sum of squared coefficients) to the loss function.

### Mathematical Formula

**Objective Function:**

```
Minimize: SSE + λ Σβⱼ²
```

Expanded:
```
Minimize: Σ(yᵢ - ŷᵢ)² + λ(β₁² + β₂² + β₃² + ... + βₚ²)
```

**Note:** Typically, β₀ (intercept) is NOT penalized.

### Alternative Form (Constrained Optimization)

```
Minimize: Σ(yᵢ - ŷᵢ)²
Subject to: Σβⱼ² ≤ t
```

Where t is a threshold (related to λ).

### How Ridge Works

1. **Penalizes large coefficients quadratically**
2. **Shrinks all coefficients toward zero** (but never exactly to zero)
3. **Proportional shrinkage**: Larger coefficients shrink more

### Closed-Form Solution

Unlike OLS, Ridge has a modified solution:

**OLS Solution:**
```
β = (XᵀX)⁻¹Xᵀy
```

**Ridge Solution:**
```
β_ridge = (XᵀX + λI)⁻¹Xᵀy
```

Where I is the identity matrix.

**Key benefit:** XᵀX + λI is always invertible (even when XᵀX is singular).

### Example: Ridge in Action

**Original coefficients (OLS):**
```
β₁ = 50.0
β₂ = -30.0
β₃ = 20.0
```

**Ridge coefficients (λ = 10):**
```
β₁ = 35.2  (shrunk by 29.6%)
β₂ = -21.1 (shrunk by 29.7%)
β₃ = 14.1  (shrunk by 29.5%)
```

**Notice:** All coefficients shrunk proportionally, none became exactly zero.

### Properties of Ridge

#### ✅ Advantages:
1. **Handles multicollinearity**: Stabilizes coefficients when features are correlated
2. **Always has solution**: (XᵀX + λI) is always invertible
3. **Computationally efficient**: Closed-form solution exists
4. **Keeps all features**: Doesn't eliminate variables
5. **Smooth penalty**: Differentiable everywhere

#### ❌ Disadvantages:
1. **No feature selection**: All features remain in model (with small weights)
2. **Less interpretable**: All features contribute
3. **Requires scaling**: Sensitive to feature scales

### When Ridge Helps Most

**Scenario 1: Multicollinearity**
```
Features: Height_in_cm and Height_in_inches (perfectly correlated)

OLS: β₁ = 1000, β₂ = -1000 (unstable, large, opposite signs)
Ridge: β₁ = 50, β₂ = -48 (stable, smaller, manageable)
```

**Scenario 2: More Features than Observations (p > n)**
```
Observations: 50
Features: 100

OLS: Cannot solve (XᵀX not invertible)
Ridge: Works perfectly (XᵀX + λI always invertible)
```

**Scenario 3: Small Dataset**
```
Data: 30 samples, 10 features

OLS: High variance, overfits
Ridge: Reduces variance, generalizes better
```

## 5. Lasso Regression (L1 Regularization)

### Definition

Lasso (Least Absolute Shrinkage and Selection Operator) adds an **L1 penalty** (sum of absolute coefficients) to the loss function.

### Mathematical Formula

**Objective Function:**

```
Minimize: SSE + λ Σ|βⱼ|
```

Expanded:
```
Minimize: Σ(yᵢ - ŷᵢ)² + λ(|β₁| + |β₂| + |β₃| + ... + |βₚ|)
```

### Alternative Form (Constrained Optimization)

```
Minimize: Σ(yᵢ - ŷᵢ)²
Subject to: Σ|βⱼ| ≤ t
```

### How Lasso Works

1. **Penalizes coefficients linearly**
2. **Shrinks some coefficients to EXACTLY zero**
3. **Performs automatic feature selection**
4. **Creates sparse models**

### No Closed-Form Solution

Unlike Ridge, Lasso requires iterative optimization:
- Coordinate Descent
- LARS (Least Angle Regression)
- Proximal Gradient Descent

### Example: Lasso in Action

**Original coefficients (OLS):**
```
β₁ = 50.0
β₂ = -30.0
β₃ = 20.0
β₄ = 5.0
β₅ = 2.0
```

**Lasso coefficients (λ = 10):**
```
β₁ = 42.3  (shrunk)
β₂ = -22.1 (shrunk)
β₃ = 13.5  (shrunk)
β₄ = 0.0   (eliminated!)
β₅ = 0.0   (eliminated!)
```

**Notice:** Some coefficients became exactly zero → feature selection!

### Properties of Lasso

#### ✅ Advantages:
1. **Automatic feature selection**: Sets irrelevant features to zero
2. **Interpretable models**: Fewer features to explain
3. **Sparse solutions**: Only important features remain
4. **Good for high-dimensional data**: Reduces dimensionality
5. **Handles redundant features**: Picks one from correlated group

#### ❌ Disadvantages:
1. **Arbitrary selection**: Among correlated features, picks randomly
2. **Unstable with high correlation**: Different random selections
3. **Computationally expensive**: No closed-form solution
4. **Non-differentiable at zero**: Optimization is trickier
5. **Requires scaling**: Very sensitive to feature scales

### When Lasso Helps Most

**Scenario 1: Feature Selection**
```
Features: 100
Truly important: 10

OLS: Uses all 100 (overfits, hard to interpret)
Lasso: Selects ~12 most important (sparse, interpretable)
```

**Scenario 2: High-Dimensional Data**
```
Genomics: 10,000 genes, 100 patients

OLS: Impossible
Lasso: Identifies key genes (e.g., 50 genes)
```

**Scenario 3: Interpretability Required**
```
Business model: Need to explain which factors matter

OLS: "All 50 variables contribute"
Lasso: "These 8 variables drive sales"
```

## 6. Elastic Net (L1 + L2)

### Definition

Elastic Net combines both L1 (Lasso) and L2 (Ridge) penalties, getting benefits of both.

### Mathematical Formula

**Objective Function:**

```
Minimize: SSE + λ₁ Σ|βⱼ| + λ₂ Σβⱼ²
```

**Alternative parameterization:**
```
Minimize: SSE + λ[(1-α)Σβⱼ² + α Σ|βⱼ|]
```

Where:
- **λ**: Overall regularization strength
- **α**: Mix ratio (0 ≤ α ≤ 1)
  - α = 0: Pure Ridge
  - α = 1: Pure Lasso
  - 0 < α < 1: Elastic Net (typically α = 0.5)

### How Elastic Net Works

**Combines strengths of both:**
1. **From Ridge**: Handles correlated features well
2. **From Lasso**: Performs feature selection

**Visual Analogy:**
```
Ridge:       "Shrink everything gradually"
Lasso:       "Eliminate or keep"
Elastic Net: "Group similar features, eliminate unimportant ones"
```

### Example: Elastic Net in Action

**Scenario:** 3 highly correlated features (multicollinearity)

**OLS:**
```
β₁ = 100, β₂ = -95, β₃ = 90  (unstable, huge, opposite signs)
```

**Ridge:**
```
β₁ = 30, β₂ = 28, β₃ = 29  (stable, but keeps all 3)
```

**Lasso:**
```
β₁ = 85, β₂ = 0, β₃ = 0  (picks 1 randomly, eliminates others)
```

**Elastic Net:**
```
β₁ = 35, β₂ = 32, β₃ = 0  (keeps correlated group, eliminates least important)
```

### Properties of Elastic Net

#### ✅ Advantages:
1. **Best of both worlds**: Stability + feature selection
2. **Handles correlated features**: Better than Lasso
3. **Encourages grouping**: Correlated features selected/eliminated together
4. **Flexible**: Can tune toward Ridge or Lasso
5. **Robust**: Works well in various scenarios

#### ❌ Disadvantages:
1. **Two hyperparameters**: λ and α (more tuning needed)
2. **Computationally expensive**: Iterative optimization
3. **Complex interpretation**: Harder to explain than Ridge/Lasso alone

### When Elastic Net Helps Most

**Scenario 1: Grouped Variables**
```
Marketing data: 
- Email_clicks, Email_opens, Email_sent (correlated)
- Social_likes, Social_shares, Social_comments (correlated)

Elastic Net: Selects important groups, shrinks within groups
```

**Scenario 2: p > n with Correlation**
```
Genomics: 5000 genes, 100 patients, genes in pathways (correlated)

Lasso: Unstable, random selection from correlated genes
Ridge: Keeps all, no selection
Elastic Net: Selects key pathways, stable results
```

## 7. Mathematical Deep Dive

### 7.1 Why L1 Creates Sparsity (Lasso)

**Geometric Intuition:**

The constraint regions have different shapes:

**Ridge (L2) constraint:**
```
β₁² + β₂² ≤ t  (circle/sphere)

    β₂
     |
     |    ___
     |  ⁄     ＼
     | |   •   |  ← smooth circle
     |  ＼_____⁄
     +------------- β₁
```

**Lasso (L1) constraint:**
```
|β₁| + |β₂| ≤ t  (diamond/rhombus)

    β₂
     |
     |    /\
     |   /  \
     |  /  • \  ← corners at axes
     | <      >
     +------------- β₁
        \  /
         \/
```

**Key difference:** 
- Lasso's diamond has **corners on the axes**
- When SSE contours hit a corner → coefficient = 0
- Ridge's circle has **no corners** → coefficients approach but never reach 0

### 7.2 Mathematical Comparison

| Aspect | OLS | Ridge | Lasso |
|--------|-----|-------|-------|
| Objective | Σ(yᵢ-ŷᵢ)² | Σ(yᵢ-ŷᵢ)² + λΣβⱼ² | Σ(yᵢ-ŷᵢ)² + λΣ\|βⱼ\| |
| Penalty Shape | None | Smooth (quadratic) | Angular (absolute) |
| Solution | Closed-form | Closed-form | Iterative |
| Sparsity | No | No | Yes |
| Differentiable | Yes | Yes | No (at 0) |

### 7.3 Effect of λ on Coefficients

**As λ increases:**

```
λ = 0      λ = 1      λ = 10     λ = 100    λ → ∞

Ridge:
β₁ = 50.0 → 45.2 → 35.1 → 15.3 → 0.0
β₂ = 30.0 → 27.1 → 21.0 → 9.2  → 0.0
β₃ = 10.0 → 9.0  → 7.0  → 3.1  → 0.0

Lasso:
β₁ = 50.0 → 48.0 → 40.0 → 25.0 → 0.0
β₂ = 30.0 → 28.0 → 20.0 → 0.0  → 0.0
β₃ = 10.0 → 8.0  → 0.0  → 0.0  → 0.0
```

**Notice:**
- **Ridge**: All shrink gradually
- **Lasso**: Some hit zero earlier (feature selection)

### 7.4 Gradient Analysis

**Ridge Gradient:**
```
∂L/∂βⱼ = -2Σ(yᵢ - ŷᵢ)xᵢⱼ + 2λβⱼ
```
- Smooth, continuous
- Gradient proportional to βⱼ
- Larger coefficients get larger penalties

**Lasso Gradient (Subgradient):**
```
∂L/∂βⱼ = -2Σ(yᵢ - ŷᵢ)xᵢⱼ + λ·sign(βⱼ)
```
- Discontinuous at βⱼ = 0
- Fixed penalty regardless of size
- Can push small coefficients to exactly zero

### 7.5 Standardization is Critical

**Why features must be standardized:**

Imagine two features:
- x₁: Income (range: 20,000 - 200,000)
- x₂: Age (range: 20 - 80)

**Without standardization:**
```
Penalty = λ(β₁² + β₂²)

To affect y by 1 unit:
β₁ needs to be small (since x₁ is large)
β₂ needs to be large (since x₂ is small)

β₁² will naturally be smaller
β₂² will naturally be larger

Regularization unfairly penalizes β₂!
```

**With standardization:**
```
Both features have mean=0, std=1
Coefficients on equal footing
Fair penalization
```

**Standardization formula:**
```
x_scaled = (x - mean(x)) / std(x)
```

## 9. When to Use Which Method

### Decision Flowchart

```
Start: Do you have overfitting?
    |
    | No → Use OLS (standard linear regression)
    |
    | Yes
    ↓
Is interpretability critical?
    |
    | No → Consider Ridge
    |
    | Yes
    ↓
Are features highly correlated?
    |
    | No → Use Lasso
    |
    | Yes → Use Elastic Net
```

### Detailed Scenarios

#### Use **OLS** when:
✓ Small number of features (< 20)
✓ Large dataset (n >> p)
✓ Features are not correlated
✓ No overfitting observed
✓ Simple, interpretable model sufficient
✓ Features are all theoretically important

**Example:** Predicting house price from 5 key features (size, location, age, bedrooms, bathrooms) with 10,000 samples.

#### Use **Ridge** when:
✓ Many features (p is large)
✓ Features are correlated (multicollinearity)
✓ All features are potentially relevant
✓ Overfitting is occurring
✓ p > n (more features than samples)
✓ Want to keep all features but stabilize coefficients

**Example:** Predicting patient health from 100 correlated medical measurements with 200 patients.

#### Use **Lasso** when:
✓ Many features, but only few are important
✓ Need automatic feature selection
✓ Interpretability is crucial
✓ Features are not highly correlated
✓ Want a sparse model
✓ High-dimensional data (p >> n)

**Example:** Identifying key genes (from 10,000) that predict disease from 100 patient samples.

#### Use **Elastic Net** when:
✓ Features are correlated AND want feature selection
✓ p > n with multicollinearity
✓ Lasso is too unstable
✓ Need both stability and sparsity
✓ Groups of correlated features exist

**Example:** Predicting sales from 200 marketing metrics (many correlated) with 300 observations.

### Real-World Domain Guidelines

#### Finance & Economics
- **Stock prediction**: Ridge (many correlated indicators)
- **Credit scoring**: Lasso (identify key risk factors)
- **Portfolio optimization**: Ridge (correlated assets)

#### Healthcare & Biology
- **Genomics**: Lasso or Elastic Net (thousands of genes, few important)
- **Disease prediction**: Elastic Net (correlated symptoms)
- **Drug discovery**: Lasso (identify active compounds)

#### Marketing & Business
- **Customer churn**: Lasso (identify key drivers)
- **Sales forecasting**: Ridge (many historical features)
- **Pricing optimization**: Ridge (correlated market factors)

#### Technology & Engineering
- **Sensor data**: Ridge (correlated measurements)
- **Feature selection**: Lasso (identify critical sensors)
- **Signal processing**: Elastic Net (grouped frequency components)

## 10. Hyperparameter Tuning

### 10.1 The λ (Lambda) Parameter

**Effect of λ:**

```
λ = 0:      No regularization (OLS)
λ → 0:      Weak regularization
λ small:    Some shrinkage
λ moderate: Balanced trade-off (usually optimal)
λ large:    Heavy shrinkage
λ → ∞:      All coefficients → 0 (useless model)
```

### 10.2 Cross-Validation for λ

**K-Fold Cross-Validation:**

1. Split data into K folds (typically K=5 or 10)
2. For each λ value:
   - For each fold:
     - Train on K-1 folds
     - Validate on remaining fold
   - Average validation error across folds
3. Select λ with lowest average validation error

**Validation Curve:**

```
Error
  |                    
  |        \          
  |         \        ← Training Error
  |          \      
  |           \___   ← Validation Error
  |               \  (U-shape)
  |                \
  |                 ↑ Optimal λ
  +----------------------------
     0.001  0.01  0.1   1   10
                    λ
```

## Summary: Quick Reference Guide

### Mathematical Formulas

| Method | Objective Function | Penalty | Sparsity |
|--------|-------------------|---------|----------|
| **OLS** | Σ(yᵢ-ŷᵢ)² | None | No |

### Key Characteristics

|  | Ridge | Lasso | Elastic Net |
|--|-------|-------|-------------|
| **Feature selection** | ✗ | ✓ | ✓ |
| **Handles correlation** | ✓ | ✗ | ✓ |
| **Sparse solutions** | ✗ | ✓ | ✓ |
| **Computational cost** | Low | Medium | Medium |
| **Interpretability** | Medium | High | High |
| **Hyperparameters** | 1 (λ) | 1 (λ) | 2 (λ, α) |

### When to Use What

```
┌─────────────────────────────────────┐
│ Is p > n or severe overfitting?    │
│        No → Try OLS first           │
│        Yes ↓                        │
├─────────────────────────────────────┤
│ Need feature selection?             │
│        No → Ridge                   │
│        Yes ↓                        │
├─────────────────────────────────────┤
│ High multicollinearity?             │
│        No → Lasso                   │
│        Yes → Elastic Net            │
└─────────────────────────────────────┘
```

### Practical Workflow

1. **Preprocess data**
   - Handle missing values
   - **Standardize features** (critical!)
   - Split train/test

2. **Choose method**
   - Start with Ridge (safe default)
   - Try Lasso if need feature selection
   - Use Elastic Net if unsure

3. **Tune hyperparameters**
   - Use cross-validation
   - Test λ on log scale
   - Monitor training/validation curves

4. **Evaluate**
   - Test set performance
   - Coefficient magnitudes
   - Number of features (for Lasso/EN)

5. **Interpret**
   - Which features matter?
   - Are coefficients reasonable?
   - Does model generalize?
