## 1. Introduction

### What is Linear Regression?

Linear regression is a **supervised learning algorithm** used for predictive modeling. It establishes a linear relationship between:
- **Dependent Variable (y)**: The target variable we want to predict (also called response variable, output, or label)
- **Independent Variable(s) (x)**: The predictor variable(s) used to make predictions (also called features, inputs, or explanatory variables)

### Types of Linear Regression

1. **Simple Linear Regression**: One independent variable
   - Example: Predicting salary based on years of experience
   
2. **Multiple Linear Regression**: Two or more independent variables
   - Example: Predicting house price based on size, location, and age

### Why "Linear"?

The relationship between variables is modeled as a **linear equation** (straight line or hyperplane in higher dimensions).

## 2. Mathematical Foundation

### The Core Concept

Linear regression aims to find the **best-fitting line** through a set of data points by minimizing the distance between the actual data points and the predicted values.

### Notation

| Symbol | Meaning |
|--------|---------|
| n | Number of observations (data points) |
| x | Independent variable (predictor) |
| y | Actual value of dependent variable |
| ŷ | Predicted value of dependent variable (y-hat) |
| β₀ | Intercept (bias term) |
| β₁ | Slope (coefficient/weight) |
| ε | Error term (residual) |
| x̄ | Mean of x values |
| ȳ | Mean of y values |
| Σ | Summation symbol |

### The Linear Model

For each observation i:

```
yᵢ = β₀ + β₁xᵢ + εᵢ
```

Where εᵢ represents the error (difference between actual and predicted values).

## 3. Simple Linear Regression

### The Equation

The simple linear regression equation is:

```
ŷ = β₀ + β₁x
```

**Alternative notations you might see:**
- y = mx + b (slope-intercept form)
- y = a + bx
- y = w₀ + w₁x (machine learning notation)

### Components Explained

#### β₀ (Intercept/Bias)
- The value of y when x = 0
- Where the line crosses the y-axis
- Represents the baseline value

#### β₁ (Slope/Coefficient)
- The rate of change in y for a unit change in x
- If β₁ > 0: positive relationship (as x increases, y increases)
- If β₁ < 0: negative relationship (as x increases, y decreases)
- If β₁ = 0: no linear relationship

### Visual Interpretation

```
     y
     |
 400 |                    * (actual point)
     |                  /
 300 |              * /  (predicted line)
     |            /
 200 |        * /
     |      /
 100 |    /
     |  /
   0 |/_________________ x
     0   1000   2000   3000
```

The red line represents ŷ = β₀ + β₁x, and blue dots are actual data points.

## 4. The Least Squares Method

### Objective: Minimize Error

The goal is to find β₀ and β₁ that minimize the difference between actual and predicted values.

### Error (Residual)

For each data point i:

```
eᵢ = yᵢ - ŷᵢ = yᵢ - (β₀ + β₁xᵢ)
```

### Why Square the Errors?

We use **Sum of Squared Errors (SSE)** instead of sum of errors:

```
SSE = Σ(yᵢ - ŷᵢ)²
```

**Reasons:**
1. **Positive and negative errors don't cancel out**: Squaring makes all errors positive
2. **Penalizes large errors more**: An error of 10 contributes 100 to SSE, while two errors of 5 contribute only 50 total
3. **Mathematically convenient**: Makes the optimization problem differentiable

### Mathematical Derivation

To minimize SSE, we take partial derivatives with respect to β₀ and β₁ and set them to zero:

```
∂(SSE)/∂β₀ = 0
∂(SSE)/∂β₁ = 0
```

### The Normal Equations

Solving these equations gives us the **Ordinary Least Squares (OLS)** estimators:

#### Formula for Slope (β₁):

```
β₁ = Σ[(xᵢ - x̄)(yᵢ - ȳ)] / Σ(xᵢ - x̄)²
```

**Alternative form:**

```
β₁ = [n·Σ(xᵢyᵢ) - Σxᵢ·Σyᵢ] / [n·Σ(xᵢ²) - (Σxᵢ)²]
```

**Covariance form:**

```
β₁ = Cov(x,y) / Var(x)
```

**Correlation form:**

```
β₁ = r · (sᵧ / sₓ)
```

Where:
- r = correlation coefficient between x and y
- sᵧ = standard deviation of y
- sₓ = standard deviation of x

#### Formula for Intercept (β₀):

```
β₀ = ȳ - β₁·x̄
```

**Important property**: The regression line always passes through the point (x̄, ȳ).

## 5. Detailed Step-by-Step Example

### Problem Statement

A real estate company wants to predict house prices based on the size of the house. They have collected data on 5 houses.

### Dataset

| House # | Size (x) sq ft | Price (y) $1000s |
|---------|----------------|------------------|
| 1       | 1000           | 200              |
| 2       | 1500           | 250              |
| 3       | 2000           | 300              |
| 4       | 2500           | 350              |
| 5       | 3000           | 400              |

### Step 1: Calculate Means

```
x̄ = (1000 + 1500 + 2000 + 2500 + 3000) / 5
x̄ = 10000 / 5
x̄ = 2000
```

```
ȳ = (200 + 250 + 300 + 350 + 400) / 5
ȳ = 1500 / 5
ȳ = 300
```

### Step 2: Create Calculation Table

We need to calculate several components for β₁:

| i | xᵢ   | yᵢ  | (xᵢ - x̄) | (yᵢ - ȳ) | (xᵢ - x̄)(yᵢ - ȳ) | (xᵢ - x̄)² |
|---|------|-----|----------|----------|-------------------|-----------|
| 1 | 1000 | 200 | -1000    | -100     | 100,000           | 1,000,000 |
| 2 | 1500 | 250 | -500     | -50      | 25,000            | 250,000   |
| 3 | 2000 | 300 | 0        | 0        | 0                 | 0         |
| 4 | 2500 | 350 | 500      | 50       | 25,000            | 250,000   |
| 5 | 3000 | 400 | 1000     | 100      | 100,000           | 1,000,000 |
| **Σ** |      |     |          |          | **250,000**       | **2,500,000** |

### Step 3: Calculate Slope (β₁)

```
β₁ = Σ[(xᵢ - x̄)(yᵢ - ȳ)] / Σ(xᵢ - x̄)²
β₁ = 250,000 / 2,500,000
β₁ = 0.1
```

**Interpretation**: For every additional square foot, the house price increases by $100 (since 0.1 × 1000 = 100).

### Step 4: Calculate Intercept (β₀)

```
β₀ = ȳ - β₁·x̄
β₀ = 300 - 0.1(2000)
β₀ = 300 - 200
β₀ = 100
```

**Interpretation**: The base price of a house (when size = 0) is $100,000. This is theoretical since houses can't have 0 sq ft.

### Step 5: Write the Final Equation

```
ŷ = 100 + 0.1x
```

**Or in words:**
```
Price (in $1000s) = 100 + 0.1 × Size (in sq ft)
```

### Step 6: Make Predictions

#### Prediction 1: For x = 2200 sq ft
```
ŷ = 100 + 0.1(2200)
ŷ = 100 + 220
ŷ = 320
```
**Predicted price: $320,000**

#### Prediction 2: For x = 1800 sq ft
```
ŷ = 100 + 0.1(1800)
ŷ = 100 + 180
ŷ = 280
```
**Predicted price: $280,000**

#### Prediction 3: For x = 3500 sq ft
```
ŷ = 100 + 0.1(3500)
ŷ = 100 + 350
ŷ = 450
```
**Predicted price: $450,000**

### Step 7: Verify Predictions on Training Data

| House | Actual x | Actual y | Predicted ŷ | Error (eᵢ) | Squared Error |
|-------|----------|----------|-------------|-----------|---------------|
| 1     | 1000     | 200      | 100+0.1(1000)=200 | 0 | 0 |
| 2     | 1500     | 250      | 100+0.1(1500)=250 | 0 | 0 |
| 3     | 2000     | 300      | 100+0.1(2000)=300 | 0 | 0 |
| 4     | 2500     | 350      | 100+0.1(2500)=350 | 0 | 0 |
| 5     | 3000     | 400      | 100+0.1(3000)=400 | 0 | 0 |
| **Sum** |      |          |             |           | **SSE = 0** |

**Note**: SSE = 0 means perfect fit! This is rare with real data. Our example was constructed to demonstrate the concepts clearly.

## 6. Model Evaluation Metrics

### Why Evaluate?

We need metrics to assess:
1. **How well does the model fit the training data?**
2. **How well will it generalize to new data?**
3. **Is the model better than simply using the mean?**

### 6.1 Sum of Squared Errors (SSE)

Also called **Residual Sum of Squares (RSS)** or **Sum of Squared Residuals (SSR)**:

```
SSE = Σ(yᵢ - ŷᵢ)²
```

- **Lower is better**
- Depends on the scale of y
- Difficult to interpret in isolation

### 6.2 Mean Squared Error (MSE)

Average of squared errors:

```
MSE = SSE / n = (1/n) Σ(yᵢ - ŷᵢ)²
```

- **Lower is better**
- Same units as y²
- Penalizes large errors heavily

### 6.3 Root Mean Squared Error (RMSE)

Square root of MSE:

```
RMSE = √MSE = √[(1/n) Σ(yᵢ - ŷᵢ)²]
```

- **Lower is better**
- Same units as y (easier to interpret)
- For house prices: RMSE = 15 means predictions are off by about $15,000 on average

### 6.4 Mean Absolute Error (MAE)

Average of absolute errors:

```
MAE = (1/n) Σ|yᵢ - ŷᵢ|
```

- **Lower is better**
- Same units as y
- Less sensitive to outliers than MSE/RMSE

### 6.5 R² (R-squared) - Coefficient of Determination

**Most important metric for linear regression:**

```
R² = 1 - (SSE / SST)
```

Where:
```
SST = Σ(yᵢ - ȳ)² (Total Sum of Squares)
```

**Alternative formulas:**

```
R² = (SST - SSE) / SST
```

```
R² = SSR / SST
```

Where SSR = Sum of Squares due to Regression = Σ(ŷᵢ - ȳ)²

**Interpretation:**
- **Range**: 0 to 1 (can be negative for poorly fitted models)
- **R² = 1**: Perfect fit (all variance explained)
- **R² = 0.8**: 80% of variance in y is explained by x
- **R² = 0**: Model is no better than using the mean ȳ
- **Higher is better**

**For simple linear regression:**

```
R² = r²
```

Where r is the Pearson correlation coefficient between x and y.

### 6.6 Adjusted R²

Adjusts R² for the number of predictors (important for multiple regression):

```
Adjusted R² = 1 - [(1 - R²)(n - 1) / (n - p - 1)]
```

Where:
- n = number of observations
- p = number of predictors

**Purpose**: Prevents overfitting by penalizing unnecessary variables.

### Example Calculation

Using our house price data:

#### Calculate SST:
```
SST = Σ(yᵢ - ȳ)²
SST = (200-300)² + (250-300)² + (300-300)² + (350-300)² + (400-300)²
SST = 10000 + 2500 + 0 + 2500 + 10000
SST = 25000
```

#### We already know:
```
SSE = 0
```

#### Calculate R²:
```
R² = 1 - (SSE / SST)
R² = 1 - (0 / 25000)
R² = 1
```

**Interpretation**: Our model explains 100% of the variance in house prices!

## 7. Multiple Linear Regression

### Concept

When you have multiple independent variables, the equation becomes:

```
ŷ = β₀ + β₁x₁ + β₂x₂ + β₃x₃ + ... + βₚxₚ
```

Where:
- p = number of predictors
- β₀ = intercept
- β₁, β₂, ..., βₚ = coefficients for each predictor

### Example: House Price Prediction

Predicting house price using multiple features:

```
Price = β₀ + β₁(Size) + β₂(Bedrooms) + β₃(Age) + β₄(Distance_to_City)
```

### Matrix Notation

With multiple variables, calculations are done using matrices:

**Data matrix X:**
```
X = [1  x₁₁  x₁₂  ...  x₁ₚ]
    [1  x₂₁  x₂₂  ...  x₂ₚ]
    [⋮   ⋮    ⋮   ...   ⋮ ]
    [1  xₙ₁  xₙ₂  ...  xₙₚ]
```

**Target vector y:**
```
y = [y₁]
    [y₂]
    [⋮ ]
    [yₙ]
```

**Coefficient vector β:**
```
β = [β₀]
    [β₁]
    [β₂]
    [⋮ ]
    [βₚ]
```

### The Normal Equation (Matrix Form)

```
β = (XᵀX)⁻¹Xᵀy
```

Where:
- Xᵀ = transpose of X
- (XᵀX)⁻¹ = inverse of XᵀX

### Example with Two Predictors

**Data:**

| Size (x₁) | Bedrooms (x₂) | Price (y) |
|-----------|---------------|-----------|
| 1000      | 2             | 200       |
| 1500      | 3             | 250       |
| 2000      | 3             | 300       |
| 2500      | 4             | 350       |
| 3000      | 4             | 400       |

**Model:**
```
Price = β₀ + β₁(Size) + β₂(Bedrooms)
```

**Matrix setup:**

```
X = [1  1000  2]        y = [200]
    [1  1500  3]            [250]
    [1  2000  3]            [300]
    [1  2500  4]            [350]
    [1  3000  4]            [400]
```

Solving β = (XᵀX)⁻¹Xᵀy gives approximate coefficients (this requires matrix computation):

```
β₀ ≈ 50
β₁ ≈ 0.08
β₂ ≈ 30
```

**Final equation:**
```
Price = 50 + 0.08(Size) + 30(Bedrooms)
```

**Interpretation:**
- **β₁ = 0.08**: Each additional sq ft adds $80, holding bedrooms constant
- **β₂ = 30**: Each additional bedroom adds $30,000, holding size constant
- **β₀ = 50**: Base price is $50,000

### Multicollinearity

**Problem**: When independent variables are highly correlated with each other.

**Example**: Including both "size in sq ft" and "size in sq meters" would cause multicollinearity.

**Detection**: 
- Variance Inflation Factor (VIF) > 10 indicates problematic multicollinearity

**Solutions**:
- Remove one of the correlated variables
- Combine them into a single variable
- Use regularization techniques (Ridge, Lasso)

## 8. Assumptions of Linear Regression

Linear regression relies on several key assumptions. Violating these can lead to unreliable predictions.

### 8.1 Linearity

**Assumption**: The relationship between x and y is linear.

**Check**: 
- Scatter plot of x vs y should show a linear pattern
- Residual plot (residuals vs fitted values) should show no pattern

**What if violated?**
- Transform variables (log, square root, polynomial)
- Use polynomial regression
- Try non-linear models

### 8.2 Independence

**Assumption**: Observations are independent of each other.

**Common violations**:
- Time series data (consecutive observations are related)
- Clustered data (students within schools)
- Spatial data (nearby locations are similar)

**Check**:
- Durbin-Watson test for autocorrelation
- Plot residuals over time

**Solutions**:
- Time series models (ARIMA, SARIMA)
- Mixed-effects models
- Generalized Estimating Equations (GEE)

### 8.3 Homoscedasticity

**Assumption**: Constant variance of errors across all levels of x.

**Heteroscedasticity**: When variance of errors changes with x.

**Check**:
- Residual plot: residuals should form a horizontal band
- Breusch-Pagan test
- White test

**Visual example**:
```
Homoscedastic (Good):        Heteroscedastic (Bad):
Residuals                    Residuals
    |  •  •  •  •                |  •
    | • • • • •                  | • •
  0 |___________              0 |_____
    | • • • • •                  | • •  •
    |  •  •  •  •                |  •  •  •
    +-----------                 +-------------
```

**Solutions**:
- Transform y (log, square root)
- Weighted Least Squares (WLS)
- Robust standard errors

### 8.4 Normality of Residuals

**Assumption**: Errors (residuals) are normally distributed.

**Check**:
- Histogram of residuals
- Q-Q plot (quantile-quantile plot)
- Shapiro-Wilk test
- Kolmogorov-Smirnov test

**Important note**: This assumption is about the **residuals**, not the raw data.

**Solutions**:
- Transform y
- Use robust regression methods
- Increase sample size (Central Limit Theorem helps)

### 8.5 No Multicollinearity (Multiple Regression)

**Assumption**: Independent variables are not highly correlated with each other.

**Check**:
- Correlation matrix
- Variance Inflation Factor (VIF)

**Solutions**:
- Remove correlated variables
- Principal Component Analysis (PCA)
- Ridge or Lasso regression

### 8.6 No Influential Outliers

**Outliers**: Data points that don't follow the pattern.

**Types**:
- **High leverage**: Unusual x values
- **High residual**: Unusual y values
- **Influential**: Significantly affects the regression line

**Detection**:
- Cook's distance > 1 suggests influential point
- Leverage > 2(p+1)/n suggests high leverage
- Standardized residuals > 3 suggest outliers

**Solutions**:
- Investigate: data entry error? Valid extreme case?
- Robust regression (RANSAC, Huber regression)
- Transform variables
- Remove if justified

## 9. Advanced Concepts

### 9.1 Gradient Descent (Alternative to Normal Equation)

An iterative optimization algorithm to find β₀ and β₁.

#### Algorithm:

1. **Initialize** β₀ and β₁ randomly (e.g., both = 0)
2. **Compute predictions**: ŷᵢ = β₀ + β₁xᵢ
3. **Compute cost** (MSE):
   ```
   J(β₀, β₁) = (1/2n) Σ(yᵢ - ŷᵢ)²
   ```
4. **Update parameters** using gradients:
   ```
   β₀ := β₀ - α · (1/n) Σ(ŷᵢ - yᵢ)
   β₁ := β₁ - α · (1/n) Σ(ŷᵢ - yᵢ) · xᵢ
   ```
   Where α is the learning rate
5. **Repeat** steps 2-4 until convergence

#### When to use?
- Very large datasets (millions of rows)
- Many features (normal equation becomes computationally expensive)
- Online learning (updating model with new data)

### 9.2 Polynomial Regression

Fits non-linear relationships by adding polynomial terms:

```
ŷ = β₀ + β₁x + β₂x² + β₃x³ + ...
```

**Example**: Quadratic (degree 2)
```
ŷ = β₀ + β₁x + β₂x²
```

**Still linear regression!** It's linear in the parameters β, even though the relationship with x is non-linear.

**Caution**: High-degree polynomials can cause **overfitting**.

### 9.3 Regularization

Techniques to prevent overfitting by penalizing large coefficients.

#### Ridge Regression (L2 Regularization)

Minimizes:
```
SSE + λ Σβⱼ²
```

- Shrinks coefficients toward zero
- Doesn't eliminate variables
- λ controls regularization strength

#### Lasso Regression (L1 Regularization)

Minimizes:
```
SSE + λ Σ|βⱼ|
```

- Can shrink coefficients to exactly zero
- Performs feature selection
- Creates sparse models

#### Elastic Net

Combines Ridge and Lasso:
```
SSE + λ₁ Σ|βⱼ| + λ₂ Σβⱼ²
```

### 9.4 Feature Scaling

**Problem**: Features with different scales can bias the model.

**Example**: 
- Size: 1000-3000 sq ft
- Age: 1-50 years

**Solutions**:

#### Standardization (Z-score normalization):
```
x_scaled = (x - μ) / σ
```
Result: mean = 0, std = 1

#### Min-Max Normalization:
```
x_scaled = (x - x_min) / (x_max - x_min)
```
Result: range = [0, 1]

**When needed?**
- Gradient descent (faster convergence)
- Regularization (fair penalization)
- Comparing coefficient magnitudes

### 9.5 Confidence Intervals and Hypothesis Testing

#### Confidence Interval for β₁

```
β₁ ± t_{α/2, n-2} · SE(β₁)
```

Where:
```
SE(β₁) = √[MSE / Σ(xᵢ - x̄)²]
```

**Example**: 95% CI for slope in house price model:
```
β₁ = 0.1 ± 1.96 × SE(β₁)
```

If CI = [0.08, 0.12], we're 95% confident the true slope is between 0.08 and 0.12.

#### Hypothesis Test for β₁

**Null hypothesis**: H₀: β₁ = 0 (no relationship)
**Alternative**: H₁: β₁ ≠ 0 (relationship exists)

**Test statistic**:
```
t = β₁ / SE(β₁)
```

**Decision**: 
- If |t| > critical value, reject H₀
- If p-value < α (e.g., 0.05), reject H₀

**Interpretation**: We have evidence that x significantly predicts y.

### 9.6 Prediction Intervals

**Difference from confidence intervals**:
- **Confidence interval**: Uncertainty about the mean response
- **Prediction interval**: Uncertainty about an individual prediction

**Formula**:
```
ŷ ± t_{α/2, n-2} · √[MSE · (1 + 1/n + (x - x̄)² / Σ(xᵢ - x̄)²)]
```

**Prediction intervals are always wider** than confidence intervals.

## Practice Problems

### Problem 1: Basic Calculation

Given the data:

| x | y |
|---|---|
| 1 | 2 |
| 2 | 4 |
| 3 | 5 |
| 4 | 4 |
| 5 | 5 |

Calculate β₀, β₁, and R².

<details>
<summary>Solution</summary>

**Step 1: Calculate means**
- x̄ = (1+2+3+4+5)/5 = 3
- ȳ = (2+4+5+4+5)/5 = 4

**Step 2: Calculate β₁**

| xᵢ | yᵢ | (xᵢ-x̄) | (yᵢ-ȳ) | (xᵢ-x̄)(yᵢ-ȳ) | (xᵢ-x̄)² |
|----|----|---------|---------|--------------  |---------|
| 1  | 2  | -2      | -2      | 4             | 4       |
| 2  | 4  | -1      | 0       | 0             | 1       |
| 3  | 5  | 0       | 1       | 0             | 0       |
| 4  | 4  | 1       | 0       | 0             | 1       |
| 5  | 5  | 2       | 1       | 2             | 4       |
| Σ  |    |         |         | 6             | 10      |

β₁ = 6/10 = 0.6

**Step 3: Calculate β₀**
- β₀ = 4 - 0.6(3) = 4 - 1.8 = 2.2

**Equation: ŷ = 2.2 + 0.6x**

**Step 4: Calculate R²**
- Predictions: 2.8, 3.4, 4.0, 4.6, 5.2
- SSE = (2-2.8)² + (4-3.4)² + (5-4.0)² + (4-4.6)² + (5-5.2)² = 2.0
- SST = (2-4)² + (4-4)² + (5-4)² + (4-4)² + (5-4)² = 6
- R² = 1 - 2.0/6 = 0.67

**Answer**: β₁ = 0.6, β₀ = 2.2, R² = 0.67 (67% of variance explained)

# Basic Regression Concepts (Simple Explanation)

## 1. Linearity

**Meaning:**\
Linearity means the relationship between the input variable (X) and
output (Y) should be a straight-line relationship.

**Example:**\
Suppose we predict salary based on years of experience.

  Years of Experience   Salary
  --------------------- --------
  1                     20,000
  2                     30,000
  3                     40,000

As experience increases, salary increases in a straight-line pattern.

**Model example** Y = a + bX

Where: - Y = predicted value - X = feature - a = intercept - b = slope

------------------------------------------------------------------------

## 2. Multicollinearity

**Meaning:**\
Multicollinearity happens when **two or more input features are highly
correlated with each other**.

This makes it difficult for the regression model to understand which
feature actually influences the output.

**Example:**

  Feature 1           Feature 2
  ------------------- -----------------
  House size (sqft)   Number of rooms

If house size increases, number of rooms usually increases too.\
So both features give **almost the same information**.

**Problem:**\
The model gets confused about which variable affects the target.

------------------------------------------------------------------------

## 3. Correlation

**Meaning:**\
Correlation measures **how strongly two variables are related**.

Correlation values range from **-1 to +1**.

  Value   Meaning
  ------- -------------------------------
  +1      Perfect positive relationship
  0       No relationship
  -1      Perfect negative relationship

**Example:**

Hours studied vs Exam score

  Hours   Score
  ------- -------
  1       40
  2       50
  3       65
  4       80

As study hours increase, score increases.\
This is **positive correlation**.

------------------------------------------------------------------------

## 4. Autocorrelation

**Meaning:**\
Autocorrelation happens when **errors (residuals) are correlated with
each other over time**.

It usually occurs in **time series data**.

**Example:**

Temperature data:

  Day         Temperature
  ----------- -------------
  Monday      30
  Tuesday     31
  Wednesday   32

Today's temperature is related to yesterday's temperature.

This creates **autocorrelation in residuals**.

------------------------------------------------------------------------

## 5. No Autocorrelation

**Meaning:**\
In a good regression model, **errors should not depend on previous
errors**.

Each error should be independent.

**Example:**

Prediction errors:

  Observation   Error
  ------------- -------
  1             2
  2             -1
  3             3
  4             -2

Errors appear random with **no pattern**.

This is **no autocorrelation**.

------------------------------------------------------------------------

## 6. Homoscedasticity

**Meaning:**\
Homoscedasticity means the **variance of errors remains constant** for
all values of X.

In simple words:\
The spread of errors should stay roughly the same.

**Example:**

If we plot predicted vs error:

Small X → errors around ±2\
Large X → errors around ±2

The spread stays constant.

------------------------------------------------------------------------

## 7. Heteroscedasticity (Opposite of Homoscedasticity)

**Meaning:**\
Error variance **changes as X increases**.

Example pattern:

Small X → errors around ±2\
Large X → errors around ±10

The spread increases.

This causes problems in regression analysis.

------------------------------------------------------------------------

# Quick Summary

  Concept              Meaning
  -------------------- ------------------------------------------------
  Linearity            Relationship between X and Y should be linear
  Multicollinearity    Independent variables strongly correlated
  Correlation          Strength of relationship between two variables
  Autocorrelation      Errors correlated over time
  No Autocorrelation   Errors independent
  Homoscedasticity     Constant variance of errors
  Heteroscedasticity   Changing variance of errors


```markdown
# Regression Evaluation Metrics

When we build a regression model, we need to measure **how good the predictions are**.  
Some common metrics are:

- Mean Absolute Error (MAE)
- Mean Squared Error (MSE)
- Root Mean Squared Error (RMSE)
- R² Score (R Square)

---

# 1. Mean Absolute Error (MAE)

## Definition
Mean Absolute Error measures the **average absolute difference between actual values and predicted values**.

It does not care whether the error is positive or negative.

## Formula

MAE = (1/n) Σ |y - ŷ|

Where:
- **y** = actual value
- **ŷ** = predicted value
- **n** = number of observations

## Example

Actual house prices:
```

[200, 250, 300]

```

Predicted prices:
```

[210, 240, 310]

```

Errors:

| Actual | Predicted | Error | Absolute Error |
|------|------|------|------|
|200|210|10|10|
|250|240|-10|10|
|300|310|10|10|

MAE = (10 + 10 + 10) / 3 = **10**

## Key Characteristics
- Easy to understand
- Treats all errors **equally**
- Not sensitive to outliers

## When to Use
Use MAE when:
- You want **simple interpretation**
- Outliers should **not dominate the error**

Example:
- Demand prediction
- Sales forecasting

---

# 2. Mean Squared Error (MSE)

## Definition
MSE measures the **average of squared differences between actual and predicted values**.

Large errors get **much higher penalty**.

## Formula

MSE = (1/n) Σ (y - ŷ)²

## Example

Using the same data:

Errors:

| Actual | Predicted | Error | Squared Error |
|------|------|------|------|
|200|210|10|100|
|250|240|-10|100|
|300|310|10|100|

MSE = (100 + 100 + 100) / 3 = **100**

## Key Characteristics
- Penalizes **large errors heavily**
- Always positive
- Useful in optimization algorithms

## When to Use
Use MSE when:
- Large errors are **very bad**
- Training machine learning models

Example:
- Neural networks
- Gradient descent models

Reason:
Because **squared errors are differentiable**, which helps optimization.

---

# 3. Root Mean Squared Error (RMSE)

## Definition
RMSE is the **square root of MSE**.

It converts the error back to the **original unit of the target variable**.

## Formula

RMSE = √(MSE)

or

RMSE = √[(1/n) Σ (y - ŷ)²]

## Example

From previous example:

MSE = 100

RMSE = √100 = **10**

## Key Characteristics
- Same unit as target variable
- Penalizes large errors
- Easier to interpret than MSE

## When to Use
Use RMSE when:
- You want **interpretable error values**
- Large errors must be penalized

Example:
- Weather prediction
- Price prediction
- Stock prediction

---

# 4. R² Score (R Square)

## Definition
R² measures **how well the model explains the variance in the data**.

It tells **how much of the target variable is explained by the model**.

## Formula

R² = 1 − (SS_res / SS_tot)

Where:

SS_res = Σ (y − ŷ)²  
SS_tot = Σ (y − ȳ)²

- **y** = actual value
- **ŷ** = predicted value
- **ȳ** = mean of actual values

## Value Range

| R² Value | Meaning |
|------|------|
|1|Perfect model|
|0|Model explains nothing|
|Negative|Model worse than mean|

## Example

If R² = **0.85**

It means:

**85% of the variance in the data is explained by the model.**

---

# 5. Key Differences

| Metric | Measures | Outlier Sensitivity | Unit |
|------|------|------|------|
|MAE|Average absolute error|Low|Same as target|
|MSE|Average squared error|High|Squared unit|
|RMSE|Square root of MSE|High|Same as target|
|R²|Variance explained|Depends|Unitless|

---

# 6. Which Metric is Better?

There is **no single best metric**. It depends on the problem.

## Use MAE when
- Outliers should **not affect the model much**
- You want **simple interpretation**

Example:
Delivery time prediction.

---

## Use MSE when
- Large errors are **very dangerous**
- Training ML models with **gradient descent**

Example:
Medical prediction models.

---

## Use RMSE when
- You want **penalty for large errors**
- But still want **error in original units**

Example:
House price prediction.

---

## Use R² when
- You want to know **how well the model explains the data**

Example:
Model evaluation in research papers.

---

# 7. Simple Intuition

Think like this:

MAE → average mistake  
MSE → punish big mistakes heavily  
RMSE → punish big mistakes but easy to interpret  
R² → how much of the data the model explains

---

# 8. Example Summary

Suppose a house price model gives:

MAE = 10,000  
RMSE = 15,000  
R² = 0.85  

Meaning:
- Average error = **₹10,000**
- Large errors exist (RMSE higher)
- Model explains **85% of the data**
```
