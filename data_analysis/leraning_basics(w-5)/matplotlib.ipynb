# How to Analyze the Box Plot

A **box plot** shows the **statistical distribution of data** using five important values.

## 1. Minimum Value
- The **bottom whisker** represents the **minimum value** in the dataset.
- In this plot it is around **25**.

## 2. First Quartile (Q1)
- The **bottom of the box** represents the **25th percentile**.
- About **25% of the data lies below this value**.
- Here it is around **35**.

## 3. Median (Q2)
- The **line inside the box** represents the **median (middle value)**.
- Half of the data is **below** this value and half is **above** it.
- In this plot it is around **40**.

## 4. Third Quartile (Q3)
- The **top of the box** represents the **75th percentile**.
- About **75% of the data lies below this value**.
- Here it is around **50**.

## 5. Maximum Value
- The **top whisker** represents the **maximum value**.
- In this plot it is around **60**.

## What This Plot Tells Us
- Most data lies between **35 and 50** (inside the box).
- The **median value is around 40**.
- The overall data range is **25 to 60**.
- There are **no outliers shown** in this plot.

```python
data = [10, 12, 13, 15, 18, 20, 22, 25, 30, 100]

plt.boxplot(data)
plt.title("Box Plot with Outliers")
plt.show()
```

# Why the Box Plot Does Not Draw a Whisker Line at 100

In a box plot, the **whiskers do not always go to the maximum value**.

They only extend up to a limit based on the **Interquartile Range (IQR)** rule.

## Outlier Rule

Outliers are values outside this range:

Lower Limit  
= Q1 − 1.5 × IQR  

Upper Limit  
= Q3 + 1.5 × IQR  

Where:
- **Q1** = First quartile (25%)
- **Q3** = Third quartile (75%)
- **IQR** = Q3 − Q1

## What Happens in Your Example

Dataset:
25 + 1.5 × 12
= 25 + 18
= 43


Any value **greater than 43 is an outlier**.

Since **100 > 43**, it is treated as an **outlier**.

## Result in the Plot

- The **whisker stops at the largest normal value** (around 30).
- **100 is shown as a separate dot**, not as the whisker line.

## Simple Meaning

The whisker shows the **normal data range**, while points like **100 are marked separately as outliers**.

# How to Analyze a Violin Plot

A **violin plot** shows the **distribution and density of the data**.  
It combines features of a **box plot and a density plot**.

## 1. Shape of the Violin
- The **width of the violin** shows how dense the data is.
- **Wider area → more data points**
- **Narrow area → fewer data points**

## 2. Center Line
- The **vertical line inside** shows the **range of the data** from minimum to maximum.

## 3. Top and Bottom Ends
- **Top end** → maximum value.
- **Bottom end** → minimum value.

## 4. Density Interpretation
- If the violin is **wide near the middle**, many values are around the mean.
- If the violin is **wide at top or bottom**, more values are concentrated there.

## 5. What This Plot Shows
For your dataset (`np.random.normal(0,1,100)`):

- Data follows a **normal distribution**.
- Most values are around **0**.
- Values gradually decrease as they move toward **-3 or +2**.
- The middle area is wider, meaning **more data points are near the center**.

## Simple Meaning
A violin plot helps us see:
- **Where most data points occur**
- **How data is distributed**
- **Whether the distribution is symmetric or skewed**

```python
#stem plot
x = np.linspace(0, 2*np.pi, 10)
y = np.cos(x)
plt.stem(x, y)
plt.title("Stem Plot Example")
plt.show()
```

# How to Analyze a Stem Plot

A **stem plot** shows **discrete data points** using vertical lines (stems) from a baseline.

## 1. X-axis
- Represents the **input values**.
- In this example it shows values from **0 to 2π**.

## 2. Y-axis
- Represents the **output values** of the function.
- Here it shows **cos(x) values**.

## 3. Stems (Vertical Lines)
- Each vertical line connects the **baseline (0)** to the **data value**.
- It helps clearly see the value of each data point.

## 4. Markers (Dots)
- The dot at the top of each stem represents the **actual value of y**.

## 5. Pattern in This Plot
Since the function used is **cos(x)**:

- At **x = 0**, cos(x) ≈ **1** (maximum)
- Around **x ≈ π/2**, cos(x) decreases toward **0**
- Around **x ≈ π**, cos(x) becomes **negative**
- Around **x ≈ 3π/2**, values reach near **-1**
- Toward **2π**, the values increase back toward **1**

## Simple Meaning
The stem plot clearly shows **each discrete value of the cosine function**, making it easy to see the **rise and fall pattern of the cosine wave**.

# How to Analyze a Step Plot

A **step plot** shows data that **changes in steps instead of smooth curves**.  
The value stays constant for some range and then **suddenly jumps to another value**.

## 1. X-axis
- Represents the **input values or intervals**.
- In this plot it ranges roughly from **20 to 80**.

## 2. Y-axis
- Represents the **value of the variable** at each interval.

## 3. Step Shape
- The horizontal line shows the **value remains constant** for a period.
- The vertical jump shows **sudden change in value**.

## 4. What This Plot Shows
- From **x ≈ 20**, value starts around **10** then jumps to **40**.
- Between **30–70**, the value repeatedly switches between **30 and 100**.
- Near **80**, the value decreases again.

## 5. Why Step Plots Are Used
Step plots are useful when data changes **discretely at specific points**, such as:
- **Digital signals**
- **Stock price changes at intervals**
- **System states**
- **Piecewise functions**

## Simple Meaning
A step plot shows **values that remain constant for some time and then change suddenly**, forming a staircase-like pattern.

# How to Analyze a Contour Plot

A **contour plot** represents a **3D surface on a 2D graph** using contour lines.  
Each line connects points having the **same value (Z value)**.

## 1. X-axis and Y-axis
- Represent the **input variables**.
- Here they range from **-3 to 3**.

## 2. Contour Lines
- Each circular line shows **points with equal Z values**.
- These are similar to **height lines on a map**.

## 3. Color Gradient
- The **color bar on the side** shows the value of **Z**.
- Dark colors represent **lower values**.
- Brighter colors represent **higher values**.

## 4. Pattern in This Plot
Since  
`Z = √(X² + Y²)`

- The value depends on **distance from the center (0,0)**.
- At the **center**, Z is smallest.
- As we move **away from the center**, Z increases.

This is why the contour lines form **concentric circles**.

## 5. What the Plot Shows
- Center area → **small values**
- Outer circles → **larger values**
- Each circle represents **equal distance from the center**.

## Simple Meaning
A contour plot helps visualize **how a variable changes across two dimensions**, where each line represents **points with the same value**.

```python
import numpy as np

x = np.array([1, 2, 3])
y = np.array([4, 5])

X, Y = np.meshgrid(x, y)

print("X:\n", X)
print("Y:\n", Y)
```

# np.meshgrid() – Simple Explanation

`np.meshgrid()` is used to **create coordinate grids from two 1D arrays**.

It helps generate **X and Y coordinate matrices** so we can evaluate a function for every (x, y)

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.cos(Y)

plt.contour(X, Y, Z, cmap='coolwarm')
plt.title("Contour Plot Example")
plt.colorbar()

plt.show()
```

```python
#polar plot
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(2*theta))
plt.polar(theta, r)
plt.title("Polar Plot Example")
plt.show()
```

# How to Analyze This Polar Plot

## What is a Polar Plot
A **polar plot** represents data using:
- **Angle (θ)** → direction
- **Radius (r)** → distance from the center

Instead of **(x, y)** coordinates, it uses **(θ, r)**.

---

# Components of the Plot

## 1. Center
The middle point is **radius = 0**.

## 2. Circular Rings
These represent **distance from the center**.

Example:
- inner ring → r = 0.2
- outer ring → r = 1

## 3. Angle Labels
Around the circle we see angles:
0°, 45°, 90°, 135°, 180°, etc.

These represent **θ values**.

---

# Function Used

r = |sin(2θ)|

Meaning:
- Radius depends on the **angle θ**.

---

# Pattern in the Plot

Because of **sin(2θ)**:

- The value repeats **4 times in one full circle**.
- This creates **4 petal shapes**.

When:
- sin(2θ) = 0 → r = 0 → curve touches center
- sin(2θ) = 1 → r = 1 → maximum distance

---

# How to Read the Plot

1. Choose an **angle θ**.
2. Compute **r = |sin(2θ)|**.
3. Move **r distance from the center in that direction**.

That point is plotted.

---

# Simple Meaning

This polar plot shows how the **radius changes with angle**, forming a **four-petal flower shape** due to the periodic sine function.

# 3D Plots

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') 
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title("3D Surface Plot")
plt.show()
```

# Simple Explanation of Each Line

## 1. Create X Values
`X = np.linspace(-5, 5, 30)`

- Generates **30 evenly spaced numbers** from **-5 to 5**.
- These values represent **positions along the X-axis**.

Example (simplified):
-5, -4.6, -4.2 ... 4.6, 5

---

## 2. Create Y Values
`Y = np.linspace(-5, 5, 30)`

- Generates **30 evenly spaced numbers** from **-5 to 5**.
- These represent **positions along the Y-axis**.

---

## 3. Create Coordinate Grid
`X, Y = np.meshgrid(X, Y)`

This creates a **2D grid of coordinates**.

Meaning:
- Every **X value pairs with every Y value**.
- This forms **all possible (x, y) coordinate points**.

Example grid:

X matrix

-5 -4 -3

-5 -4 -3

-5 -4 -3



Y matrix

-5 -5 -5

-4 -4 -4

-3 -3 -3


Each pair gives coordinates like:
(-5,-5), (-4,-5), (-3,-5), etc.

---

## 4. Compute Z Values
`Z = np.sin(np.sqrt(X**2 + Y**2))`

This calculates a **value for every (x,y) point**.

Steps:

### Step 1
`X**2 + Y**2`

Computes the **distance squared from the center (0,0)**.

---

### Step 2
`np.sqrt(X**2 + Y**2)`

Calculates the **distance from the center**.

Example:
- Point (0,0) → distance = 0
- Point (3,4) → distance = 5

---

### Step 3
`np.sin(...)`

Applies the **sine function to that distance**.

This creates a **wave pattern spreading from the center**.

---

# What This Produces

The function creates **circular waves** around the center.

Like **ripples in water when you drop a stone**.

Values:
- Oscillate between **-1 and 1**
- Form **rings expanding outward**.

---

# Simple Meaning

For every point (x,y):
1. Find its **distance from the center**.
2. Apply the **sine function** to that distance.
3. Store the result in **Z**.

This creates a **wave-like surface used for contour or surface plots**.

2. Compute its **distance from the center (0,0)**.
3. Apply the **sin() function** to that distance.

---

# Shape of the Surface

You can see:

- **Peaks (hills)** where sin value is positive
- **Valleys (dips)** where sin value is negative
- **Circular ripple pattern**

It looks like **waves spreading outward from the center**.

---

# Simple Interpretation

This plot shows how the value **Z changes for every (X,Y) point**, forming a **3D wave surface with circular ripples**.

```python
#3d line plot
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t
 
ax.plot3D(x, y, z, 'purple')
ax.set_title("3D Line Plot")
plt.show()
```

```python
# 3D scatter plot

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
 
ax.scatter(x, y, z, color='red', marker='o')
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
# 3D wireframe plot
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_title("3D Wireframe Plot")
plt.show()
```

```python
# 3D contour plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour3D(X, Y, Z, 50, cmap='plasma')
ax.set_title("3D Contour Plot")
plt.show()
```

```python
#3d bar plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = np.zeros(5)
dx = np.ones(5)
dy = np.ones(5)
dz = [5, 2, 7, 3, 6]
 
ax.bar3d(x, y, z, dx, dy, dz, color='teal')
ax.set_title("3D Bar Plot")
plt.show()
```

```python
# 3d contour plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))
 
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')
fig.colorbar(surf)
ax.set_title("3D Surface with Colormap")
plt.show()
 
```

```python
# 3D Sphere Surface
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='skyblue')
ax.set_title("3D Sphere Surface")
plt.show()
```

```python
# 3D Surface + Wireframe
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7, cmap='viridis')
ax.plot_wireframe(X, Y, Z, color='black', linewidth=0.5)
ax.set_title("3D Surface + Wireframe")
plt.show()
 
```

```python
# 3D Contour Projection on Planes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour(X, Y, Z, zdir='z', offset=-2, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='x', offset=-5, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='y', offset=5, cmap='coolwarm')
 
ax.set_title("3D Contour Projection on Planes")
plt.show()
```
